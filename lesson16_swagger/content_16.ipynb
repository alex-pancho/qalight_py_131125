# –û—Ä–≥–∞–Ω—ñ–∑–∞—Ü—ñ—è —Ç–µ—Å—Ç—ñ–≤ API –º–µ—Ç–æ–¥–æ–º Decoupled Testing

–ü—ñ–¥—Ö—ñ–¥, –∫–æ–ª–∏ —Ç–µ—Å—Ç–∏ —ñ –æ–ø–∏—Å API —Ä–æ–∑–¥—ñ–ª–µ–Ω—ñ, –Ω–∞–∑–∏–≤–∞—î—Ç—å—Å—è —Ä–æ–∑–ø–æ–¥—ñ–ª–µ–Ω–∏–º —Ç–µ—Å—Ç—É–≤–∞–Ω–Ω—è–º (**Decoupled Testing**) –∞–±–æ —Ç–µ—Å—Ç—É–≤–∞–Ω–Ω—è–º –∑ —Ä–æ–∑–¥—ñ–ª–µ–Ω–Ω—è–º –æ–±–æ–≤'—è–∑–∫—ñ–≤ (**Separation of Concerns**). –¶–µ–π –ø—ñ–¥—Ö—ñ–¥ –º–∞—î –∫—ñ–ª—å–∫–∞ –ø–µ—Ä–µ–≤–∞–≥ —ñ –º–æ–∂–µ –±—É—Ç–∏ –∫–æ—Ä–∏—Å–Ω–∏–º —É –≤–µ–ª–∏–∫–∏—Ö –ø—Ä–æ–µ–∫—Ç–∞—Ö –∞–±–æ —É –≤–∏–ø–∞–¥–∫—É —Å–ø—ñ–ª—å–Ω–æ—ó —Ä–æ–±–æ—Ç–∏ —Ä—ñ–∑–Ω–∏—Ö –∫–æ–º–∞–Ω–¥:

```python
import logging

def pytest_configure(config):
    # –ù–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è —Ä—ñ–≤–Ω—è –ª–æ–≥—É–≤–∞–Ω–Ω—è (DEBUG, INFO, WARNING, ERROR, CRITICAL)
    config.option.log_cli_level = 'INFO'

    # –í—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω—è —Ñ–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—è –ª–æ–≥—ñ–≤
    logging.basicConfig(
        level=logging.INFO,
        format='%(asctime)s - %(levelname)s - %(message)s'
    )

    # –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –æ–±'—î–∫—Ç–∞ FileHandler –¥–ª—è –∑–±–µ—Ä–µ–∂–µ–Ω–Ω—è –ª–æ–≥—ñ–≤ —É —Ñ–∞–π–ª
    file_handler = logging.FileHandler('pytest.log')
    file_handler.setLevel(logging.INFO)
    file_handler.setFormatter(logging.Formatter('%(asctime)s - %(levelname)s - %(message)s'))

    # –î–æ–¥–∞–≤–∞–Ω–Ω—è –æ–±'—î–∫—Ç–∞ FileHandler –¥–æ –ª–æ–≥–µ—Ä–∞
    logging.getLogger().addHandler(file_handler)
```
