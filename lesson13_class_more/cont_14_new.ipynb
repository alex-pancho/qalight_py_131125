### 1.2 __new__() - –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –æ–±'—î–∫—Ç–∞

## 2. –ú–∞–≥—ñ—á–Ω—ñ –º–µ—Ç–æ–¥–∏ (Magic/Dunder Methods)

–ú–∞–≥—ñ—á–Ω—ñ –º–µ—Ç–æ–¥–∏ - —Ü–µ —Å–ø–µ—Ü—ñ–∞–ª—å–Ω—ñ –º–µ—Ç–æ–¥–∏, —è–∫—ñ –∑–º—ñ–Ω—é—é—Ç—å –ø–æ–≤–µ–¥—ñ–Ω–∫—É –æ–±'—î–∫—Ç—ñ–≤ –ø—Ä–∏ –≤–∑–∞—î–º–æ–¥—ñ—ó –∑ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞–º–∏ —Ç–∞ –≤–±—É–¥–æ–≤–∞–Ω–∏–º–∏ —Ñ—É–Ω–∫—Ü—ñ—è–º–∏.

### 2.1 –ú–µ—Ç–æ–¥–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—è –æ–±'—î–∫—Ç–∞

| –ú–µ—Ç–æ–¥ | –û–ø–∏—Å | –í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è |
|-------|------|--------------|
| `__str__()` | –ö—Ä–∞—Å–∏–≤–∞ –≤—ñ–¥–ø–æ–≤—ñ–¥—å –¥–ª—è –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á–∞ | `print()`, `str()` |
| `__repr__()` | –û—Ñ—ñ—Ü—ñ–π–Ω–∞ –ø—Ä–µ–¥—Å—Ç–∞–≤–∞ –¥–ª—è —Ä–æ–∑—Ä–æ–±–Ω–∏–∫–∞ | `repr()`, –∫–æ–Ω—Å–æ–ª—å |
| `__len__()` | –í–∏–∑–Ω–∞—á–∞—î –¥–æ–≤–∂–∏–Ω—É –æ–±'—î–∫—Ç–∞ | `len()` |

```python
class CarParams:
    def __init__(self, make, model):
        self.make = make
        self.__model = model
    
    def __repr__(self):
        return f"CarParams('{self.make}', '{self.__model}')"
    
    def __str__(self):
        return f"The {self.make} {self.__model} object"
    
    def __len__(self):
        return 100

my_car = CarParams("Toyota", "Camry")
print(my_car)        # __str__: The Toyota Camry object
print(repr(my_car))  # __repr__: CarParams('Toyota', 'Camry')
print(len(my_car))   # __len__: 100
```

### 2.2 –ê—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω—ñ –º–∞–≥—ñ—á–Ω—ñ –º–µ—Ç–æ–¥–∏

–î–æ–∑–≤–æ–ª—è—é—Ç—å –≤–∏–∑–Ω–∞—á–∏—Ç–∏ –ø–æ–≤–µ–¥—ñ–Ω–∫—É –æ–ø–µ—Ä–∞—Ç–æ—Ä—ñ–≤ `+`, `-`, `*`, `/` —Ç–∞ —ñ–Ω—à–∏—Ö.

```python
class Money:
    def __init__(self, amount: float) -> None:
        self.__amount = amount

    def __str__(self):
        return f"{self.__amount:.2f}"

    def __add__(self, other):
        if not isinstance(other, (float, int, Money)):
            raise TypeError(f"int or float expected, {type(other)} got")
        if isinstance(other, Money):
            return Money(self.__amount + other._Money__amount)
        else:
            return Money(self.__amount + other)

    def __sub__(self, other):
        if not isinstance(other, (float, int, Money)):
            raise TypeError(f"int or float expected, {type(other)} got")
        if isinstance(other, Money):
            return Money(self.__amount - other._Money__amount)
        else:
            return Money(self.__amount - other)
    
    @property
    def amount(self):
        return self.__amount

# –í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è
money1 = Money(100)
money2 = Money(50)
result = money1 + money2  # –í–∏–∫–ª–∏–∫–∞—î __add__
print(result)  # 150.00
```

**–¢–∞–±–ª–∏—Ü—è –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω–∏—Ö –º–µ—Ç–æ–¥—ñ–≤:**

| –ú–µ—Ç–æ–¥ | –û–ø–µ—Ä–∞—Ç–æ—Ä | –ü—Ä–∏–∫–ª–∞–¥ |
|-------|----------|--------|
| `__add__` | `+` | `a + b` |
| `__sub__` | `-` | `a - b` |
| `__mul__` | `*` | `a * b` |
| `__truediv__` | `/` | `a / b` |
| `__floordiv__` | `//` | `a // b` |
| `__mod__` | `%` | `a % b` |
| `__pow__` | `**` | `a ** b` |

### 2.3 –ú–µ—Ç–æ–¥–∏ –ø–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è

```python
class Money:
    def __eq__(self, other):  # ==
        return self.__amount == other.__amount
    
    def __ne__(self, other):  # !=
        return self.__amount != other.__amount
    
    def __lt__(self, other):  # <
        return self.__amount < other.__amount
    
    def __le__(self, other):  # <=
        return self.__amount <= other.__amount
    
    def __gt__(self, other):  # >
        return self.__amount > other.__amount
    
    def __ge__(self, other):  # >=
        return self.__amount >= other.__amount

m1 = Money(100)
m2 = Money(50)
print(m1 > m2)   # True
print(m1 == m2)  # False
```

## 3. –Ü–Ω–∫–∞–ø—Å—É–ª—è—Ü—ñ—è (Encapsulation)

### 3.2 Properties - –í–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ

–î–æ–∑–≤–æ–ª—è—é—Ç—å –æ–±–º–µ–∂–∏—Ç–∏ –¥–æ—Å—Ç—É–ø –¥–æ –∞—Ç—Ä–∏–±—É—Ç—ñ–≤ —ñ –¥–æ–¥–∞—Ç–∏ –≤–∞–ª—ñ–¥–∞—Ü—ñ—é.

```python
class CarParams:
    def __init__(self, make, model):
        self.make = make
        self.__model = model
    
    @property  # –ì–µ—Ç–µ—Ä (–æ—Ç—Ä–∏–º–∞–Ω–Ω—è –∑–Ω–∞—á–µ–Ω–Ω—è)
    def model(self):
        return self.__model
    
    @model.setter  # –°–µ—Ç–µ—Ä (–≤—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω—è –∑–Ω–∞—á–µ–Ω–Ω—è)
    def model(self, value):
        if value == "":
            raise ValueError("Cant use empty value")
        else:
            self.__model = value

car = CarParams("Toyota", "Camry")
print(car.model)      # Camry (–≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –≥–µ—Ç–µ—Ä)
car.model = "Corolla" # –í—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω—è —á–µ—Ä–µ–∑ —Å–µ—Ç–µ—Ä
# car.model = ""      # ValueError: Cant use empty value
```

### 3.3 –ü—Ä–∏–∫–ª–∞–¥ –∑ –≤–∞–ª—ñ–¥–∞—Ü—ñ—î—é

```python
class Person:
    def __init__(self, age: int = 0) -> None:
        self.__age = age
        self.__name = ""

    @property
    def age(self):
        return self.__age
    
    @age.setter
    def age(self, years: int):
        if not isinstance(years, int):
            raise ValueError("age must be int")
        if years >= self.__age:
            self.__age = years
    
    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, new_name: str):
        if not isinstance(new_name, str):
            raise ValueError("name must be str")
        if len(new_name) >= len(self.__name):
            self.__name = new_name

person = Person(25)
person.age = 26  # OK
# person.age = "25"  # ValueError
```

## 4. –ü–æ–ª—ñ–º–æ—Ä—Ñ—ñ–∑–º (Polymorphism)

–¶–µ –∑–¥–∞—Ç–Ω—ñ—Å—Ç—å –æ–±'—î–∫—Ç—ñ–≤ —Ä—ñ–∑–Ω–∏—Ö –∫–ª–∞—Å—ñ–≤ —Ä–µ–∞–≥—É–≤–∞—Ç–∏ –Ω–∞ –æ–¥–Ω–µ –π —Ç–µ —Å–∞–º–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –ø–æ-—Ä—ñ–∑–Ω–æ–º—É.

### 4.1 –ü–æ–ª—ñ–º–æ—Ä—Ñ—ñ–∑–º —á–µ—Ä–µ–∑ –Ω–∞—Å–ª—ñ–¥—É–≤–∞–Ω–Ω—è

```python
class Money:
    def __init__(self, amount: float) -> None:
        self.__amount = amount
    
    @property
    def amount(self):
        return self.__amount

class UAH(Money):
    def __init__(self, amount: float) -> None:
        super().__init__(amount)

    def __str__(self):
        return f"{self.amount:.2f} –≥—Ä–Ω"


class USD(Money):
    def __init__(self, amount: float) -> None:
        super().__init__(amount)

    def __str__(self):
        return f"${self.amount:.2f}"

# –ü–æ–ª—ñ–º–æ—Ä—Ñ—ñ–∑–º
currencies = [UAH(100), USD(50), UAH(200)]
for currency in currencies:
    print(currency)  # –ö–æ–∂–µ–Ω –∫–ª–∞—Å –≤–∏–≤–æ–¥–∏—Ç—å –ø–æ-—Å–≤–æ—î–º—É
# –í–∏—Ö—ñ–¥:
# 100.00 –≥—Ä–Ω
# $50.00
# 200.00 –≥—Ä–Ω
```

### 4.2 –ü–æ–ª—ñ–º–æ—Ä—Ñ—ñ–∑–º —á–µ—Ä–µ–∑ —É—Ç–∏—Ü—å-—Ç–∏–ø—ñ–∑–∞—Ü—ñ—é (Duck Typing)

> "–Ø–∫—â–æ —Ü–µ –ø–æ—Ö–æ–¥–∏—Ç—å —è–∫ –∫–∞—á–∫–∞, –ø–ª–∞–≤–∞—î —è–∫ –∫–∞—á–∫–∞ —ñ –∫—Ä—è–∫–∞—î —è–∫ –∫–∞—á–∫–∞, —Ç–æ —Ü–µ, –º–∞–±—É—Ç—å, –∫–∞—á–∫–∞"

```python
class Dog:
    def speak(self):
        return "–ì–∞–≤!"

class Cat:
    def speak(self):
        return "–ú—è—É!"

class Cow:
    def speak(self):
        return "–ú—É!"

def make_animal_speak(animal):
    print(animal.speak())

animals = [Dog(), Cat(), Cow()]
for animal in animals:
    make_animal_speak(animal)
# –í–∏—Ö—ñ–¥:
# –ì–∞–≤!
# –ú—è—É!
# –ú—É!
```

### 4.3 –ü—Ä–∞–∫—Ç–∏—á–Ω–∏–π –ø—Ä–∏–∫–ª–∞–¥: –ö–æ–Ω–≤–µ—Ä—Ç–µ—Ä –≤–∞–ª—é—Ç

```python
class ForEx:
    def __init__(self, exchange_rate_by, exchange_rate_sell):
        self.exchange_rate_by = exchange_rate_by
        self.exchange_rate_sell = exchange_rate_sell
    
    def convert(self, cur_amount):
        if isinstance(cur_amount, UAH):
            usd_amount = cur_amount.amount / self.exchange_rate_sell
            return USD(usd_amount)
        elif isinstance(cur_amount, USD):
            uah_amount = cur_amount.amount * self.exchange_rate_by
            return UAH(uah_amount)
        else:
            raise TypeError("UAH or USD expected")

privatbank = ForEx(42.05, 43.57)

usd_cash = USD(100)
uah_result = privatbank.convert(usd_cash)
print(uah_result)  # –í–∏–≤–æ–¥–∏—Ç—å —Å—É–º—É –≤ –≥—Ä–∏–≤–Ω—è—Ö

uah_cash = UAH(3000)
usd_result = privatbank.convert(uah_cash)
print(usd_result)  # –í–∏–≤–æ–¥–∏—Ç—å —Å—É–º—É –≤ –¥–æ–ª–∞—Ä–∞—Ö
```

## 5. @classmethod —Ç–∞ @staticmethod

### 5.1 @staticmethod - –°—Ç–∞—Ç–∏—á–Ω—ñ –º–µ—Ç–æ–¥–∏

–ú–µ—Ç–æ–¥–∏, —è–∫—ñ –Ω–µ –ø–æ—Ç—Ä–µ–±—É—é—Ç—å –¥–æ—Å—Ç—É–ø –¥–æ –µ–∫–∑–µ–º–ø–ª—è—Ä–∞ –∫–ª–∞—Å—É –∞–±–æ —Å–∞–º–æ–≥–æ –∫–ª–∞—Å—É.

```python
class Math:
    @staticmethod
    def add(a, b):
        return a + b
    
    @staticmethod
    def multiply(a, b):
        return a * b

result1 = Math.add(5, 3)          # –ú–æ–∂–Ω–∞ –≤–∏–∫–ª–∏–∫–∞—Ç–∏ –±–µ–∑ —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –æ–±'—î–∫—Ç–∞
result2 = Math.multiply(5, 3)
print(result1)  # 8
print(result2)  # 15

# –¢–∞–∫–æ–∂ –º–æ–∂–Ω–∞ –≤–∏–∫–ª–∏–∫–∞—Ç–∏ –≤—ñ–¥ –æ–±'—î–∫—Ç–∞
obj = Math()
result3 = obj.add(10, 20)  # 30
```

### 5.2 @classmethod - –ú–µ—Ç–æ–¥–∏ –∫–ª–∞—Å—É

–ú–µ—Ç–æ–¥–∏, —è–∫—ñ –º–∞—é—Ç—å –¥–æ—Å—Ç—É–ø –¥–æ –∫–ª–∞—Å—É —á–µ—Ä–µ–∑ –ø–∞—Ä–∞–º–µ—Ç—Ä `cls`. –ß–∞—Å—Ç–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—Ç—å—Å—è —è–∫ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ñ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä–∏.

```python
class Money:
    def __init__(self, amount: float):
        self.amount = amount
    
    @classmethod
    def from_string(cls, money_str):
        """–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∏–π –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä –∑ —Ä—è–¥–∫–∞"""
        amount = float(money_str.replace("$", "").replace("–≥—Ä–Ω", ""))
        return cls(amount)
    
    @classmethod
    def zero(cls):
        """–ü–æ–≤–µ—Ä—Ç–∞—î –æ–±'—î–∫—Ç –∑ –Ω—É–ª—å–æ–≤–æ—é —Å—É–º–æ—é"""
        return cls(0)

money1 = Money(100)
money2 = Money.from_string("$50.00")
money3 = Money.zero()

print(money1.amount)  # 100
print(money2.amount)  # 50.0
print(money3.amount)  # 0
```

### 5.3 –ü–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è –º–µ—Ç–æ–¥—ñ–≤

| –¢–∏–ø –º–µ—Ç–æ–¥—É | –ü–∞—Ä–∞–º–µ—Ç—Ä | –î–æ—Å—Ç—É–ø –¥–æ –¥–∞–Ω–∏—Ö | –í–∏–∫–ª–∏–∫ |
|-----------|----------|----------------|--------|
| –ó–≤–∏—á–∞–π–Ω–∏–π –º–µ—Ç–æ–¥ | `self` | –î–æ –µ–∫–∑–µ–º–ø–ª—è—Ä–∞ —ñ –∫–ª–∞—Å—É | `obj.method()` |
| @classmethod | `cls` | –¢—ñ–ª—å–∫–∏ –¥–æ –∫–ª–∞—Å—É | `ClassName.method()` –∞–±–æ `obj.method()` |
| @staticmethod | ‚Äî | –ù–µ –º–∞—î –¥–æ—Å—Ç—É–ø—É | `ClassName.method()` –∞–±–æ `obj.method()` |

```python
class Example:
    class_variable = "–ö–ª–∞—Å—Å"
    
    def __init__(self, value):
        self.instance_variable = value
    
    def instance_method(self):
        """–î–æ—Å—Ç—É–ø –¥–æ –µ–∫–∑–µ–º–ø–ª—è—Ä–∞ (self) —ñ –∫–ª–∞—Å—É"""
        print(self.instance_variable)
        print(self.class_variable)
    
    @classmethod
    def class_method(cls):
        """–î–æ—Å—Ç—É–ø —Ç—ñ–ª—å–∫–∏ –¥–æ –∫–ª–∞—Å—É"""
        print(cls.class_variable)
    
    @staticmethod
    def static_method():
        """–ù–µ –º–∞—î –¥–æ—Å—Ç—É–ø—É –Ω—ñ –¥–æ —á–æ–≥–æ"""
        print("–ü—Ä–æ—Å—Ç–æ —Ñ—É–Ω–∫—Ü—ñ—è –≤ –∫–ª–∞—Å—ñ")

obj = Example("–∑–Ω–∞—á–µ–Ω–Ω—è")
obj.instance_method()      # –í–∏–≤–æ–¥–∏—Ç—å "–∑–Ω–∞—á–µ–Ω–Ω—è" —ñ "–ö–ª–∞—Å—Å"
Example.class_method()     # –í–∏–≤–æ–¥–∏—Ç—å "–ö–ª–∞—Å—Å"
Example.static_method()    # –í–∏–≤–æ–¥–∏—Ç—å "–ü—Ä–æ—Å—Ç–æ —Ñ—É–Ω–∫—Ü—ñ—è –≤ –∫–ª–∞—Å—ñ"
```

### 6.1 –ó–∞–≤–¥–∞–Ω–Ω—è 1: –ö–ª–∞—Å Bank Account

–ù–∞–ø–∏—à—ñ—Ç—å –∫–ª–∞—Å `BankAccount` –∑:
- –ü—Ä–∏–≤–∞—Ç–Ω–∏–º–∏ –∞—Ç—Ä–∏–±—É—Ç–∞–º–∏: `account_number`, `balance`, `owner`
- Property –¥–ª—è –±–µ–∑–ø–µ—á–Ω–æ–≥–æ –¥–æ—Å—Ç—É–ø—É –¥–æ `balance`
- –ú–µ—Ç–æ–¥–∏: `deposit()`, `withdraw()`, `transfer()`
- –ú–∞–≥—ñ—á–Ω–∏–π –º–µ—Ç–æ–¥ `__str__()` –¥–ª—è –≤–∏–≤–µ–¥–µ–Ω–Ω—è —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—ó

```python
class BankAccount:
    def __init__(self, owner: str, balance: float = 0):
        self.__account_number = id(self)  # –£–Ω—ñ–∫–∞–ª—å–Ω–∏–π –Ω–æ–º–µ—Ä
        self.__balance = balance
        self.__owner = owner
    
    @property
    def balance(self):
        return self.__balance
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f"–ü–æ–ø–æ–≤–Ω–µ–Ω–æ –Ω–∞ {amount}"
        raise ValueError("–°—É–º–∞ –º–∞—î –±—É—Ç–∏ –¥–æ–¥–∞—Ç–Ω–æ—é")
    
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            return f"–ó–Ω—è—Ç–æ {amount}"
        raise ValueError("–ù–µ–¥–æ—Å—Ç–∞—Ç–Ω—å–æ –∫–æ—à—Ç—ñ–≤")
    
    def transfer(self, other_account, amount):
        self.withdraw(amount)
        other_account.deposit(amount)
    
    def __str__(self):
        return f"–†–∞—Ö—É–Ω–æ–∫ {self.__owner}: {self.__balance:.2f}"

acc1 = BankAccount("–Ü–≤–∞–Ω", 1000)
acc2 = BankAccount("–ú–∞—Ä—ñ—è", 500)
acc1.transfer(acc2, 100)
print(acc1)  # –†–∞—Ö—É–Ω–æ–∫ –Ü–≤–∞–Ω: 900.00
print(acc2)  # –†–∞—Ö—É–Ω–æ–∫ –ú–∞—Ä—ñ—è: 600.00
```

### 6.2 –ó–∞–≤–¥–∞–Ω–Ω—è 2: –ö–ª–∞—Å Vector

–ù–∞–ø–∏—à—ñ—Ç—å –∫–ª–∞—Å `Vector` –∑:
- –ê—Ç—Ä–∏–±—É—Ç–∞–º–∏ `x`, `y`
- –ú–∞–≥—ñ—á–Ω–∏–º–∏ –º–µ—Ç–æ–¥–∞–º–∏: `__add__()`, `__sub__()`, `__mul__()`, `__len__()`
- Property –¥–ª—è –¥–æ–≤–∂–∏–Ω–∏ –≤–µ–∫—Ç–æ—Ä–∞
- –ú–µ—Ç–æ–¥–∏: `dot_product()`, `normalize()`

```python
import math

class Vector:
    def __init__(self, x: float, y: float):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)
    
    def __mul__(self, scalar):
        return Vector(self.x * scalar, self.y * scalar)
    
    def __len__(self):
        return math.sqrt(self.x**2 + self.y**2)
    
    @property
    def length(self):
        return len(self)
    
    def dot_product(self, other):
        return self.x * other.x + self.y * other.y
    
    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(3, 4)
v2 = Vector(1, 2)
print(v1 + v2)       # Vector(4, 6)
print(v1 - v2)       # Vector(2, 2)
print(v1 * 2)        # Vector(6, 8)
print(len(v1))       # 5.0
print(v1.dot_product(v2))  # 11
```

## –î–æ–¥–∞—Ç–∫–æ–≤–∏–π –º–∞—Ç–µ—Ä—ñ–∞–ª

### –ú–∞–≥—ñ—á–Ω—ñ –º–µ—Ç–æ–¥–∏ –∫–æ–Ω—Ç–µ–π–Ω–µ—Ä—ñ–≤

–Ø–∫—â–æ –≤–∏ —Ö–æ—á–µ—Ç–µ, —â–æ–± –≤–∞—à –∫–ª–∞—Å –º–∞–≤ –ø–æ–≤–µ–¥—ñ–Ω–∫—É —Å–ø–∏—Å–∫—É –∞–±–æ —Å–ª–æ–≤–Ω–∏–∫–∞:

```python
class CustomList:
    def __init__(self):
        self.items = []
    
    def __getitem__(self, index):
        return self.items[index]
    
    def __setitem__(self, index, value):
        self.items[index] = value
    
    def __delitem__(self, index):
        del self.items[index]
    
    def __len__(self):
        return len(self.items)
    
    def __contains__(self, item):
        return item in self.items
    
    def __iter__(self):
        return iter(self.items)
    
    def append(self, item):
        self.items.append(item)

cl = CustomList()
cl.append(1)
cl.append(2)
print(cl[0])      # 1
print(len(cl))    # 2
print(1 in cl)    # True
```

### –ö–æ–Ω—Ç–µ–∫—Å—Ç–Ω—ñ –º–µ–Ω–µ–¥–∂–µ—Ä–∏

```python
class FileManager:
    def __init__(self, filename, mode):
        self.filename = filename
        self.mode = mode
        self.file = None
    
    def __enter__(self):
        self.file = open(self.filename, self.mode)
        return self.file
    
    def __exit__(self, exc_type, exc_val, exc_tb):
        if self.file:
            self.file.close()

# –í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è
with FileManager("test.txt", "w") as f:
    f.write("Hello, World!")
```
