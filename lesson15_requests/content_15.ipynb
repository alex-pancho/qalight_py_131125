```python
import ssl
from urllib.request import urlopen, Request

# –í–∏–º–∫–Ω–µ–Ω–Ω—è –ø–µ—Ä–µ–≤—ñ—Ä–∫–∏ —Å–µ—Ä—Ç–∏—Ñ—ñ–∫–∞—Ç—ñ–≤ (–Ω–µ —Ä–µ–∫–æ–º–µ–Ω–¥—É—î—Ç—å—Å—è –≤ –ø—Ä–æ–¥–∞–∫—à–Ω—ñ)
context = ssl.create_default_context()
context.check_hostname = False
context.verify_mode = ssl.CERT_NONE

url = "https://www.example.com"

try:
    # –í–∫–∞–∑–∞–Ω–Ω—è –∫–æ–Ω—Ç–µ–∫—Å—Ç—É SSL –ø—Ä–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—ñ urllib
    response = urlopen(Request(url), context=context)
    print(response.read())
except Exception as e:
    print(f'–ü–æ–º–∏–ª–∫–∞: {e}')
```

```python
import requests

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Å–µ—Å—ñ—ó
session = requests.Session()

# –í—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω—è cookies –¥–ª—è —Å–µ—Å—ñ—ó
cookies = {'user_id': '12345', 'session_id': 'abcdef'}
session.cookies.update(cookies)

# –í–∏–∫–æ–Ω–∞–Ω–Ω—è –∑–∞–ø–∏—Ç—ñ–≤ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Å–µ—Å—ñ—ó
response1 = session.get('https://example.com/page1')
response2 = session.get('https://example.com/page2')
```

–û—Ç–∂–µ, **`requests.Session`** –¥–æ–∑–≤–æ–ª—è—î –∑–±–µ—Ä—ñ–≥–∞—Ç–∏ —Ç–∞ –æ–±–º—ñ–Ω—é–≤–∞—Ç–∏—Å—è –¥–∞–Ω–∏–º–∏ —Å–µ—Å—ñ—ó –º—ñ–∂ —Ä—ñ–∑–Ω–∏–º–∏ –∑–∞–ø–∏—Ç–∞–º–∏, —â–æ —Ä–æ–±–∏—Ç—å –≤–∑–∞—î–º–æ–¥—ñ—é –∑ –≤–µ–±-—Å–∞–π—Ç–∞–º–∏ –±—ñ–ª—å—à –µ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—é —Ç–∞ –∑—Ä—É—á–Ω–æ—é.

**HTTPAdapter**

–ö–ª–∞—Å **`requests.adapters.HTTPAdapter`** –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –Ω–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è –∞–¥–∞–ø—Ç–µ—Ä—ñ–≤, —è–∫—ñ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—Ç—å—Å—è –¥–ª—è –≤–∏–∫–æ–Ω–∞–Ω–Ω—è HTTP-–∑–∞–ø–∏—Ç—ñ–≤ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∏ **`requests`** –≤ Python. 

–ü—Ä–∏–∫–ª–∞–¥ –π–æ–≥–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è —Ä–∞–∑–æ–º –∑ **`requests.Session`**:

- **–ù–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è –∫—ñ–ª—å–∫–æ—Å—Ç—ñ —Å–ø—Ä–æ–±:**
–í—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω—è –∫—ñ–ª—å–∫–æ—Å—Ç—ñ —Å–ø—Ä–æ–± –ø–æ–≤—Ç–æ—Ä–∏—Ç–∏ HTTP-–∑–∞–ø–∏—Ç —É –≤–∏–ø–∞–¥–∫—É –Ω–µ–≤–¥–∞–ª–æ–≥–æ –ø—ñ–¥–∫–ª—é—á–µ–Ω–Ω—è –∞–±–æ —ñ–Ω—à–∏—Ö –ø–æ–º–∏–ª–æ–∫.
    
    ```python
    import requests
    from requests.adapters import HTTPAdapter
    
    # –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Å–µ—Å—ñ—ó
    session = requests.Session()
    
    # –ù–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è –∞–¥–∞–ø—Ç–µ—Ä–∞ –∑ –∫—ñ–ª—å–∫—ñ—Å—Ç—é —Å–ø—Ä–æ–± 3
    adapter = HTTPAdapter(max_retries=3)
    
    # –î–æ–¥–∞–≤–∞–Ω–Ω—è –∞–¥–∞–ø—Ç–µ—Ä–∞ –¥–æ —Å–µ—Å—ñ—ó
    session.mount('http://', adapter)
    session.mount('https://', adapter)
    
    # –í–∏–∫–æ–Ω–∞–Ω–Ω—è –∑–∞–ø–∏—Ç—É –∑ –∞–≤—Ç–æ–º–∞—Ç–∏—á–Ω–æ—é –ø–æ–≤—Ç–æ—Ä–Ω–æ—é —Å–ø—Ä–æ–±–æ—é
    response = session.get('https://example.com')
    ```
    

### –ê—É—Ç–µ–Ω—Ç–∏—Ñ—ñ–∫–∞—Ü—ñ—è

–£ –±—ñ–±–ª—ñ–æ—Ç–µ—Ü—ñ **`requests`** –≤ Python —î –∫—ñ–ª—å–∫–∞ —Å–ø–æ—Å–æ–±—ñ–≤ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –∞—É—Ç–µ–Ω—Ç–∏—Ñ—ñ–∫–∞—Ü—ñ—ó –ø—Ä–∏ –≤–∑–∞—î–º–æ–¥—ñ—ó –∑ –≤–µ–±-—Å–µ—Ä–≤–µ—Ä–∞–º–∏. –û—Å—å –¥–µ–∫—ñ–ª—å–∫–∞ —Å–ø–æ—Å–æ–±—ñ–≤ –Ω–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è –∞—É—Ç–µ–Ω—Ç–∏—Ñ—ñ–∫–∞—Ü—ñ—ó:

    ```python
    import requests
    
    response = requests.get('https://api.example.com', auth=('username', 'password'))
    ```

### –í—ñ–¥–ø—Ä–∞–≤–∫–∞ —Ç–∞ —Å–∫–∞—á—É–≤–∞–Ω–Ω—è —Ñ–∞–π–ªi–≤

–©–æ–± –≤–∏–∫–æ–Ω–∞—Ç–∏ –≤—ñ–¥–ø—Ä–∞–≤–∫–∞ —Ñ–∞–π–ª—É (file upload) –∞–±–æ —Å–∫–∞—á–∞—Ç–∏ —Ñ–∞–π–ª (file download) –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∏ **`requests`** –≤ Python, –≤–∞–º –ø–æ—Ç—Ä—ñ–±–Ω–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ –º–µ—Ç–æ–¥–∏ **`post()`** –∞–±–æ **`get()`** –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ. –û—Å—å —è–∫ —Ü–µ –º–æ–∂–Ω–∞ –∑—Ä–æ–±–∏—Ç–∏:

- **File Upload (–í—ñ–¥–ø—Ä–∞–≤–∫–∞ —Ñ–∞–π–ª—É):**

```python
import requests

# –í—ñ–¥–∫—Ä–∏—Ç—Ç—è —Ñ–∞–π–ª—É –¥–ª—è –∑–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è
with open('file_to_upload.txt', 'rb') as file:
    files = {'file': file}

    # –í–∏–∫–æ–Ω–∞–Ω–Ω—è POST-–∑–∞–ø–∏—Ç—É –∑ —Ñ–∞–π–ª–æ–º
    response = requests.post('https://example.com/upload', files=files)

# –û–±—Ä–æ–±–∫–∞ –≤—ñ–¥–ø–æ–≤—ñ–¥—ñ
print(response.text)
```

- **File Download (–°–∫–∞—á—É–≤–∞–Ω–Ω—è —Ñ–∞–π–ª—É):**

```python
import requests

# –í–∏–∫–æ–Ω–∞–Ω–Ω—è GET-–∑–∞–ø–∏—Ç—É –¥–ª—è —Å–∫–∞—á—É–≤–∞–Ω–Ω—è —Ñ–∞–π–ª—É
response = requests.get('https://example.com/download')

# –ó–∞–ø–∏—Å —Ñ–∞–π–ª—É –Ω–∞ –¥–∏—Å–∫
with open('downloaded_file.txt', 'wb') as file:
    file.write(response.content)
```

–£ –¥—Ä—É–≥–æ–º—É –ø—Ä–∏–∫–ª–∞–¥—ñ, –≤–∏ –≤–∏–∫–æ–Ω—É—î—Ç–µ GET-–∑–∞–ø–∏—Ç –¥–ª—è –æ—Ç—Ä–∏–º–∞–Ω–Ω—è –≤–º—ñ—Å—Ç—É —Ñ–∞–π–ª—É –∑ –≤–µ–±-—Å–µ—Ä–≤–µ—Ä–∞. –ü—ñ—Å–ª—è –æ—Ç—Ä–∏–º–∞–Ω–Ω—è –≤—ñ–¥–ø–æ–≤—ñ–¥—ñ, –≤–∏ –∑–∞–ø–∏—Å—É—î—Ç–µ –æ—Ç—Ä–∏–º–∞–Ω–∏–π –≤–º—ñ—Å—Ç —É —Ñ–∞–π–ª **`downloaded_file.txt`** –Ω–∞ –≤–∞—à–æ–º—É –ª–æ–∫–∞–ª—å–Ω–æ–º—É –¥–∏—Å–∫—É.

### –†–æ–±–æ—Ç–∞ –∑ —Å–∞–º–æ–ø—ñ–¥–ø–∏—Å–∞–Ω–∏–º–∏ —Å–µ—Ä—Ç–∏—Ñ—ñ–∫–∞—Ç–∞–º–∏

–î–ª—è —Ä–æ–±–æ—Ç–∏ –∑ —Å–∞–º–æ–ø—ñ–¥–ø–∏—Å–∞–Ω–∏–º–∏ —Å–µ—Ä—Ç–∏—Ñ—ñ–∫–∞—Ç–∞–º–∏  –≤–∞–º –Ω–µ–æ–±—Ö—ñ–¥–Ω–æ –ø–µ—Ä–µ–¥–∞—Ç–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä **`verify`** –∑—ñ –∑–Ω–∞—á–µ–Ω–Ω—è–º **`False`** –ø—Ä–∏ –≤–∏–∫–ª–∏–∫—É –º–µ—Ç–æ–¥—É **`requests.get()`** –∞–±–æ **`requests.post()`**. –¶–µ –¥–æ–∑–≤–æ–ª–∏—Ç—å —ñ–≥–Ω–æ—Ä—É–≤–∞—Ç–∏ –ø–µ—Ä–µ–≤—ñ—Ä–∫—É —Å–µ—Ä—Ç–∏—Ñ—ñ–∫–∞—Ç—ñ–≤ –ø—Ä–∏ –≤–∑–∞—î–º–æ–¥—ñ—ó –∑ —Å–µ—Ä–≤–µ—Ä–æ–º.

```python
import requests

response = requests.get('https://your-server.com', verify=False)
print(response.text)
```
