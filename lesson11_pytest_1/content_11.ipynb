## 5.1. –ú–æ–¥—É–ª—å –∑–Ω–∏–∂–æ–∫

**–§–∞–π–ª `bank/discount.py`:**

```python
def calculate_discount(price, category):
    """
    –ù–∞—Ä–∞—Ö–æ–≤—É—î –∑–Ω–∏–∂–∫—É –∑–∞–ª–µ–∂–Ω–æ –≤—ñ–¥ –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó –∫–ª—ñ—î–Ω—Ç–∞.
    
    Args:
        price (float): –ü–æ—á–∞—Ç–∫–æ–≤–∞ —Ü—ñ–Ω–∞ —Ç–æ–≤–∞—Ä—É
        category (str): –ö–∞—Ç–µ–≥–æ—Ä—ñ—è –∫–ª—ñ—î–Ω—Ç–∞ ('VIP', 'Student' –∞–±–æ —ñ–Ω—à–µ)
    
    Returns:
        float: –¶—ñ–Ω–∞ –ø—ñ—Å–ª—è –∑–Ω–∏–∂–∫–∏
    
    Raises:
        ValueError: –Ø–∫—â–æ —Ü—ñ–Ω–∞ –≤—ñ–¥'—î–º–Ω–∞
        TypeError: –Ø–∫—â–æ —Ü—ñ–Ω–∞ –Ω–µ —î —á–∏—Å–ª–æ–º
    
    –ó–Ω–∏–∂–∫–∏:
        - VIP: 20% (–º–Ω–æ–∂–∏–º–æ –Ω–∞ 0.8)
        - Student: 10% (–º–Ω–æ–∂–∏–º–æ –Ω–∞ 0.9)
        - –Ü–Ω—à—ñ: 0% (–±–µ–∑ –∑–º—ñ–Ω)
    """
    # –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ —Ç–∏–ø—É
    if not isinstance(price, (int, float)):
        raise TypeError("–¶—ñ–Ω–∞ –º–∞—î –±—É—Ç–∏ —á–∏—Å–ª–æ–º")
    
    # –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ –∑–Ω–∞—á–µ–Ω–Ω—è
    if price < 0:
        raise ValueError("–¶—ñ–Ω–∞ –Ω–µ –º–æ–∂–µ –±—É—Ç–∏ –≤—ñ–¥'—î–º–Ω–æ—é")
    
    # –ó–∞—Å—Ç–æ—Å—É–≤–∞–Ω–Ω—è –∑–Ω–∏–∂–∫–∏
    if category == "VIP":
        return price * 0.8  # 20% –∑–Ω–∏–∂–∫–∞
    elif category == "Student":
        return price * 0.9  # 10% –∑–Ω–∏–∂–∫–∞
    else:
        return price  # –ë–µ–∑ –∑–Ω–∏–∂–∫–∏


def calculate_bulk_discount(price, quantity):
    """
    –ù–∞—Ä–∞—Ö–æ–≤—É—î –∑–Ω–∏–∂–∫—É –∑–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ç–æ–≤–∞—Ä—É.
    
    Args:
        price (float): –¶—ñ–Ω–∞ –æ–¥–Ω–æ–≥–æ —Ç–æ–≤–∞—Ä—É
        quantity (int): –ö—ñ–ª—å–∫—ñ—Å—Ç—å —Ç–æ–≤–∞—Ä—ñ–≤
    
    Returns:
        float: –ó–∞–≥–∞–ª—å–Ω–∞ –≤–∞—Ä—Ç—ñ—Å—Ç—å –∑—ñ –∑–Ω–∏–∂–∫–æ—é
    
    –ó–Ω–∏–∂–∫–∏ –∑–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å:
        - 10-49 —à—Ç: 5% –∑–Ω–∏–∂–∫–∞
        - 50-99 —à—Ç: 10% –∑–Ω–∏–∂–∫–∞
        - 100+ —à—Ç: 15% –∑–Ω–∏–∂–∫–∞
    """
    if price < 0 or quantity < 0:
        raise ValueError("–¶—ñ–Ω–∞ —Ç–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –º–∞—é—Ç—å –±—É—Ç–∏ –¥–æ–¥–∞—Ç–Ω–∏–º–∏")
    
    total = price * quantity
    
    if quantity >= 100:
        return total * 0.85  # 15% –∑–Ω–∏–∂–∫–∞
    elif quantity >= 50:
        return total * 0.90  # 10% –∑–Ω–∏–∂–∫–∞
    elif quantity >= 10:
        return total * 0.95  # 5% –∑–Ω–∏–∂–∫–∞
    else:
        return total  # –ë–µ–∑ –∑–Ω–∏–∂–∫–∏
```

## 5.2. –ú–æ–¥—É–ª—å –±–∞–Ω–∫—ñ–≤—Å—å–∫–∏—Ö –æ–ø–µ—Ä–∞—Ü—ñ–π

**–§–∞–π–ª `bank/account.py`:**

```python
# –°–ª–æ–≤–Ω–∏–∫ –¥–ª—è –∑–±–µ—Ä—ñ–≥–∞–Ω–Ω—è –±–∞–ª–∞–Ω—Å—ñ–≤ —Ä–∞—Ö—É–Ω–∫—ñ–≤
# –ö–ª—é—á - —ñ–º'—è –≤–ª–∞—Å–Ω–∏–∫–∞, –∑–Ω–∞—á–µ–Ω–Ω—è - –±–∞–ª–∞–Ω—Å
accounts = {}

# –°–ª–æ–≤–Ω–∏–∫ –¥–ª—è –∑–±–µ—Ä—ñ–≥–∞–Ω–Ω—è —ñ—Å—Ç–æ—Ä—ñ—ó —Ç—Ä–∞–Ω–∑–∞–∫—Ü—ñ–π
# –ö–ª—é—á - —ñ–º'—è –≤–ª–∞—Å–Ω–∏–∫–∞, –∑–Ω–∞—á–µ–Ω–Ω—è - —Å–ø–∏—Å–æ–∫ —Ç—Ä–∞–Ω–∑–∞–∫—Ü—ñ–π
transactions_history = {}


def create_account(owner, initial_balance=0):
    """
    –°—Ç–≤–æ—Ä—é—î –Ω–æ–≤–∏–π –±–∞–Ω–∫—ñ–≤—Å—å–∫–∏–π —Ä–∞—Ö—É–Ω–æ–∫.
    
    Args:
        owner (str): –Ü–º'—è –≤–ª–∞—Å–Ω–∏–∫–∞ —Ä–∞—Ö—É–Ω–∫—É
        initial_balance (float): –ü–æ—á–∞—Ç–∫–æ–≤–∏–π –±–∞–ª–∞–Ω—Å (–∑–∞ –∑–∞–º–æ–≤—á—É–≤–∞–Ω–Ω—è–º 0)
    
    Returns:
        float: –ü–æ—á–∞—Ç–∫–æ–≤–∏–π –±–∞–ª–∞–Ω—Å —Å—Ç–≤–æ—Ä–µ–Ω–æ–≥–æ —Ä–∞—Ö—É–Ω–∫—É
    
    Raises:
        ValueError: –Ø–∫—â–æ —ñ–º'—è –ø–æ—Ä–æ–∂–Ω—î –∞–±–æ –±–∞–ª–∞–Ω—Å –≤—ñ–¥'—î–º–Ω–∏–π
    """
    if not isinstance(owner, str) or not owner.strip():
        raise ValueError("–Ü–º'—è –≤–ª–∞—Å–Ω–∏–∫–∞ –º–∞—î –±—É—Ç–∏ –Ω–µ–ø–æ—Ä–æ–∂–Ω—ñ–º —Ä—è–¥–∫–æ–º")
    
    if initial_balance < 0:
        raise ValueError("–ü–æ—á–∞—Ç–∫–æ–≤–∏–π –±–∞–ª–∞–Ω—Å –Ω–µ –º–æ–∂–µ –±—É—Ç–∏ –≤—ñ–¥'—î–º–Ω–∏–º")
    
    if owner in accounts:
        raise ValueError(f"–†–∞—Ö—É–Ω–æ–∫ –¥–ª—è {owner} –≤–∂–µ —ñ—Å–Ω—É—î")
    
    # –°—Ç–≤–æ—Ä—é—î–º–æ —Ä–∞—Ö—É–Ω–æ–∫
    accounts[owner] = initial_balance
    transactions_history[owner] = []
    
    return initial_balance


def get_balance(owner):
    """
    –ü–æ–≤–µ—Ä—Ç–∞—î –ø–æ—Ç–æ—á–Ω–∏–π –±–∞–ª–∞–Ω—Å —Ä–∞—Ö—É–Ω–∫—É.
    
    Args:
        owner (str): –Ü–º'—è –≤–ª–∞—Å–Ω–∏–∫–∞ —Ä–∞—Ö—É–Ω–∫—É
    
    Returns:
        float: –ü–æ—Ç–æ—á–Ω–∏–π –±–∞–ª–∞–Ω—Å
    
    Raises:
        ValueError: –Ø–∫—â–æ —Ä–∞—Ö—É–Ω–æ–∫ –Ω–µ —ñ—Å–Ω—É—î
    """
    if owner not in accounts:
        raise ValueError(f"–†–∞—Ö—É–Ω–æ–∫ –¥–ª—è {owner} –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ")
    
    return accounts[owner]


def deposit(owner, amount):
    """
    –ü–æ–ø–æ–≤–Ω—é—î —Ä–∞—Ö—É–Ω–æ–∫ –Ω–∞ –≤–∫–∞–∑–∞–Ω—É —Å—É–º—É.
    
    Args:
        owner (str): –Ü–º'—è –≤–ª–∞—Å–Ω–∏–∫–∞ —Ä–∞—Ö—É–Ω–∫—É
        amount (float): –°—É–º–∞ –¥–ª—è –ø–æ–ø–æ–≤–Ω–µ–Ω–Ω—è
    
    Returns:
        float: –ù–æ–≤–∏–π –±–∞–ª–∞–Ω—Å –ø—ñ—Å–ª—è –ø–æ–ø–æ–≤–Ω–µ–Ω–Ω—è
    
    Raises:
        ValueError: –Ø–∫—â–æ —Ä–∞—Ö—É–Ω–æ–∫ –Ω–µ —ñ—Å–Ω—É—î –∞–±–æ —Å—É–º–∞ –Ω–µ–∫–æ—Ä–µ–∫—Ç–Ω–∞
    """
    if owner not in accounts:
        raise ValueError(f"–†–∞—Ö—É–Ω–æ–∫ –¥–ª—è {owner} –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ")
    
    if amount <= 0:
        raise ValueError("–°—É–º–∞ –ø–æ–ø–æ–≤–Ω–µ–Ω–Ω—è –º–∞—î –±—É—Ç–∏ –¥–æ–¥–∞—Ç–Ω–æ—é")
    
    # –î–æ–¥–∞—î–º–æ –∫–æ—à—Ç–∏
    accounts[owner] += amount
    
    # –ó–∞–ø–∏—Å—É—î–º–æ —Ç—Ä–∞–Ω–∑–∞–∫—Ü—ñ—é
    transactions_history[owner].append(f"–ü–æ–ø–æ–≤–Ω–µ–Ω–Ω—è: +{amount}")
    
    return accounts[owner]


def withdraw(owner, amount):
    """
    –ó–Ω—ñ–º–∞—î –∫–æ—à—Ç–∏ –∑ —Ä–∞—Ö—É–Ω–∫—É.
    
    Args:
        owner (str): –Ü–º'—è –≤–ª–∞—Å–Ω–∏–∫–∞ —Ä–∞—Ö—É–Ω–∫—É
        amount (float): –°—É–º–∞ –¥–ª—è –∑–Ω—è—Ç—Ç—è
    
    Returns:
        float: –ù–æ–≤–∏–π –±–∞–ª–∞–Ω—Å –ø—ñ—Å–ª—è –∑–Ω—è—Ç—Ç—è
    
    Raises:
        ValueError: –Ø–∫—â–æ —Ä–∞—Ö—É–Ω–æ–∫ –Ω–µ —ñ—Å–Ω—É—î, —Å—É–º–∞ –Ω–µ–∫–æ—Ä–µ–∫—Ç–Ω–∞ 
                    –∞–±–æ –Ω–µ–¥–æ—Å—Ç–∞—Ç–Ω—å–æ –∫–æ—à—Ç—ñ–≤
    """
    if owner not in accounts:
        raise ValueError(f"–†–∞—Ö—É–Ω–æ–∫ –¥–ª—è {owner} –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ")
    
    if amount <= 0:
        raise ValueError("–°—É–º–∞ –∑–Ω—è—Ç—Ç—è –º–∞—î –±—É—Ç–∏ –¥–æ–¥–∞—Ç–Ω–æ—é")
    
    if amount > accounts[owner]:
        raise ValueError("–ù–µ–¥–æ—Å—Ç–∞—Ç–Ω—å–æ –∫–æ—à—Ç—ñ–≤ –Ω–∞ —Ä–∞—Ö—É–Ω–∫—É")
    
    # –ó–Ω—ñ–º–∞—î–º–æ –∫–æ—à—Ç–∏
    accounts[owner] -= amount
    
    # –ó–∞–ø–∏—Å—É—î–º–æ —Ç—Ä–∞–Ω–∑–∞–∫—Ü—ñ—é
    transactions_history[owner].append(f"–ó–Ω—è—Ç—Ç—è: -{amount}")
    
    return accounts[owner]


def transfer(sender, receiver, amount):
    """
    –ü–µ—Ä–µ–∫–∞–∑—É—î –∫–æ—à—Ç–∏ –∑ –æ–¥–Ω–æ–≥–æ —Ä–∞—Ö—É–Ω–∫—É –Ω–∞ —ñ–Ω—à–∏–π.
    
    Args:
        sender (str): –Ü–º'—è –≤—ñ–¥–ø—Ä–∞–≤–Ω–∏–∫–∞
        receiver (str): –Ü–º'—è –æ—Ç—Ä–∏–º—É–≤–∞—á–∞
        amount (float): –°—É–º–∞ –ø–µ—Ä–µ–∫–∞–∑—É
    
    Raises:
        ValueError: –Ø–∫—â–æ —Ä–∞—Ö—É–Ω–∫–∏ –Ω–µ —ñ—Å–Ω—É—é—Ç—å –∞–±–æ –Ω–µ–¥–æ—Å—Ç–∞—Ç–Ω—å–æ –∫–æ—à—Ç—ñ–≤
    """
    if sender not in accounts:
        raise ValueError(f"–†–∞—Ö—É–Ω–æ–∫ –≤—ñ–¥–ø—Ä–∞–≤–Ω–∏–∫–∞ {sender} –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ")
    
    if receiver not in accounts:
        raise ValueError(f"–†–∞—Ö—É–Ω–æ–∫ –æ—Ç—Ä–∏–º—É–≤–∞—á–∞ {receiver} –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ")
    
    # –ó–Ω—ñ–º–∞—î–º–æ –∫–æ—à—Ç–∏ –∑ —Ä–∞—Ö—É–Ω–∫—É –≤—ñ–¥–ø—Ä–∞–≤–Ω–∏–∫–∞
    withdraw(sender, amount)
    
    # –î–æ–¥–∞—î–º–æ –∫–æ—à—Ç–∏ –Ω–∞ —Ä–∞—Ö—É–Ω–æ–∫ –æ—Ç—Ä–∏–º—É–≤–∞—á–∞
    accounts[receiver] += amount
    
    # –ó–∞–ø–∏—Å—É—î–º–æ —Ç—Ä–∞–Ω–∑–∞–∫—Ü—ñ—ó
    transactions_history[sender].append(f"–ü–µ—Ä–µ–∫–∞–∑ {receiver}: -{amount}")
    transactions_history[receiver].append(f"–ü–µ—Ä–µ–∫–∞–∑ –≤—ñ–¥ {sender}: +{amount}")


def get_transaction_history(owner):
    """
    –ü–æ–≤–µ—Ä—Ç–∞—î —ñ—Å—Ç–æ—Ä—ñ—é —Ç—Ä–∞–Ω–∑–∞–∫—Ü—ñ–π –¥–ª—è —Ä–∞—Ö—É–Ω–∫—É.
    
    Args:
        owner (str): –Ü–º'—è –≤–ª–∞—Å–Ω–∏–∫–∞ —Ä–∞—Ö—É–Ω–∫—É
    
    Returns:
        list: –°–ø–∏—Å–æ–∫ —Ç—Ä–∞–Ω–∑–∞–∫—Ü—ñ–π
    
    Raises:
        ValueError: –Ø–∫—â–æ —Ä–∞—Ö—É–Ω–æ–∫ –Ω–µ —ñ—Å–Ω—É—î
    """
    if owner not in accounts:
        raise ValueError(f"–†–∞—Ö—É–Ω–æ–∫ –¥–ª—è {owner} –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ")
    
    return transactions_history[owner].copy()


def close_account(owner):
    """
    –ó–∞–∫—Ä–∏–≤–∞—î —Ä–∞—Ö—É–Ω–æ–∫ (–≤–∏–¥–∞–ª—è—î –∑ —Å–∏—Å—Ç–µ–º–∏).
    
    Args:
        owner (str): –Ü–º'—è –≤–ª–∞—Å–Ω–∏–∫–∞ —Ä–∞—Ö—É–Ω–∫—É
    
    Returns:
        float: –ó–∞–ª–∏—à–æ–∫ –∫–æ—à—Ç—ñ–≤ –Ω–∞ –º–æ–º–µ–Ω—Ç –∑–∞–∫—Ä–∏—Ç—Ç—è
    
    Raises:
        ValueError: –Ø–∫—â–æ —Ä–∞—Ö—É–Ω–æ–∫ –Ω–µ —ñ—Å–Ω—É—î
    """
    if owner not in accounts:
        raise ValueError(f"–†–∞—Ö—É–Ω–æ–∫ –¥–ª—è {owner} –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ")
    
    final_balance = accounts[owner]
    
    # –í–∏–¥–∞–ª—è—î–º–æ —Ä–∞—Ö—É–Ω–æ–∫ —ñ –π–æ–≥–æ —ñ—Å—Ç–æ—Ä—ñ—é
    del accounts[owner]
    del transactions_history[owner]
    
    return final_balance


def reset_all_accounts():
    """
    –û—á–∏—â—É—î –≤—Å—ñ —Ä–∞—Ö—É–Ω–∫–∏ (–¥–æ–ø–æ–º—ñ–∂–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è –¥–ª—è —Ç–µ—Å—Ç—ñ–≤).
    """
    accounts.clear()
    transactions_history.clear()
```

# 6. –ù–∞–ø–∏—Å–∞–Ω–Ω—è —Ç–µ—Å—Ç—ñ–≤: –ì—Ä–∞–Ω–∏—á–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è —Ç–∞ –∫–ª–∞—Å–∏ –µ–∫–≤—ñ–≤–∞–ª–µ–Ω—Ç–Ω–æ—Å—Ç—ñ

–î–ª—è —Ñ—É–Ω–∫—Ü—ñ—ó `calculate_discount`:

- **–ì—Ä—É–ø–∞ 1:** –ö–∞—Ç–µ–≥–æ—Ä—ñ—è "VIP" ‚Üí –∑–Ω–∏–∂–∫–∞ 20%
- **–ì—Ä—É–ø–∞ 2:** –ö–∞—Ç–µ–≥–æ—Ä—ñ—è "Student" ‚Üí –∑–Ω–∏–∂–∫–∞ 10%

**–î–æ—Å—Ç–∞—Ç–Ω—å–æ –ø—Ä–æ—Ç–µ—Å—Ç—É–≤–∞—Ç–∏ –æ–¥–∏–Ω –ø—Ä–∏–∫–ª–∞–¥ –∑ –∫–æ–∂–Ω–æ—ó –≥—Ä—É–ø–∏**, –∞ –Ω–µ –≤—Å—ñ –º–æ–∂–ª–∏–≤—ñ –≤–∞—Ä—ñ–∞–Ω—Ç–∏.

## 6.2. –ì—Ä–∞–Ω–∏—á–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è (Boundary Values)

**–Ü–¥–µ—è:** –ü–æ–º–∏–ª–∫–∏ –Ω–∞–π—á–∞—Å—Ç—ñ—à–µ –≤–∏–Ω–∏–∫–∞—é—Ç—å –Ω–∞ "–∫—Ä–∞—è—Ö" –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ–≤.

–î–ª—è —Ü—ñ–Ω–∏:
- `price = 0` ‚Üí –º—ñ–Ω—ñ–º–∞–ª—å–Ω–∞ –º–µ–∂–∞
- `price = 0.01` ‚Üí —Ç—Ä–æ—Ö–∏ –±—ñ–ª—å—à–µ –Ω—É–ª—è
- `price = -1` ‚Üí –Ω–µ–∫–æ—Ä–µ–∫—Ç–Ω–∞ –º–µ–∂–∞ (–º–∞—î –≤–∏–∫–∏–Ω—É—Ç–∏ –ø–æ–º–∏–ª–∫—É)

–î–ª—è –∫—ñ–ª—å–∫–æ—Å—Ç—ñ (`calculate_bulk_discount`):

- `quantity = 10` ‚Üí 5% –∑–Ω–∏–∂–∫–∞ (–ø–æ—á–∞—Ç–æ–∫ –¥—ñ–∞–ø–∞–∑–æ–Ω—É)
- `quantity = 49` ‚Üí 5% –∑–Ω–∏–∂–∫–∞ (–∫—ñ–Ω–µ—Ü—å –¥—ñ–∞–ø–∞–∑–æ–Ω—É)
- `quantity = 50` ‚Üí 10% –∑–Ω–∏–∂–∫–∞ (–ø–æ—á–∞—Ç–æ–∫ –Ω–æ–≤–æ–≥–æ –¥—ñ–∞–ø–∞–∑–æ–Ω—É)
## 6.3. –¢–µ—Å—Ç–∏ –¥–ª—è –º–æ–¥—É–ª—è –∑–Ω–∏–∂–æ–∫

```python

# ============================================
# –¢–µ—Å—Ç–∏ –∑–Ω—è—Ç—Ç—è –∫–æ—à—Ç—ñ–≤
# ============================================

def test_withdraw_decreases_balance():
    """–ó–Ω—è—Ç—Ç—è –∑–º–µ–Ω—à—É—î –±–∞–ª–∞–Ω—Å."""
    create_account("–í–∞—Å–∏–ª—å –Ü–≤–∞–Ω–µ–Ω–∫–æ", initial_balance=200)
    
    # –ó–Ω—ñ–º–∞—î–º–æ 50
    new_balance = withdraw("–í–∞—Å–∏–ª—å –Ü–≤–∞–Ω–µ–Ω–∫–æ", 50)
    
    assert new_balance == 150
    assert get_balance("–í–∞—Å–∏–ª—å –Ü–≤–∞–Ω–µ–Ω–∫–æ") == 150


def test_withdraw_all_balance():
    """–ó–Ω—è—Ç—Ç—è –≤—Å—å–æ–≥–æ –±–∞–ª–∞–Ω—Å—É."""
    create_account("–û–∫—Å–∞–Ω–∞ –ü–∞–≤–ª–µ–Ω–∫–æ", initial_balance=100)
    
    withdraw("–û–∫—Å–∞–Ω–∞ –ü–∞–≤–ª–µ–Ω–∫–æ", 100)
    
    assert get_balance("–û–∫—Å–∞–Ω–∞ –ü–∞–≤–ª–µ–Ω–∫–æ") == 0


def test_withdraw_insufficient_funds_raises_error():
    """–ó–Ω—è—Ç—Ç—è –±—ñ–ª—å—à–µ –±–∞–ª–∞–Ω—Å—É –º–∞—î –≤–∏–∫–∏–¥–∞—Ç–∏ –ø–æ–º–∏–ª–∫—É."""
    create_account("–Ü–≥–æ—Ä –ú–æ—Ä–æ–∑", initial_balance=50)
    
    with pytest.raises(ValueError, match="–ù–µ–¥–æ—Å—Ç–∞—Ç–Ω—å–æ –∫–æ—à—Ç—ñ–≤ –Ω–∞ —Ä–∞—Ö—É–Ω–∫—É"):
        withdraw("–Ü–≥–æ—Ä –ú–æ—Ä–æ–∑", 100)


def test_withdraw_from_empty_account_raises_error():
    """–ó–Ω—è—Ç—Ç—è –∑ –ø–æ—Ä–æ–∂–Ω—å–æ–≥–æ —Ä–∞—Ö—É–Ω–∫—É –º–∞—î –≤–∏–∫–∏–¥–∞—Ç–∏ –ø–æ–º–∏–ª–∫—É."""
    create_account("–û–ª–µ–∫—Å–∞–Ω–¥—Ä –ö–æ–≤–∞–ª—å—á—É–∫")
    
    with pytest.raises(ValueError):
        withdraw("–û–ª–µ–∫—Å–∞–Ω–¥—Ä –ö–æ–≤–∞–ª—å—á—É–∫", 10)


def test_withdraw_negative_raises_error():
    """–í—ñ–¥'—î–º–Ω–µ –∑–Ω—è—Ç—Ç—è –º–∞—î –≤–∏–∫–∏–¥–∞—Ç–∏ –ø–æ–º–∏–ª–∫—É."""
    create_account("–ö–∞—Ç–µ—Ä–∏–Ω–∞ Sav—á—É–∫", initial_balance=100)
    
    with pytest.raises(ValueError):
        withdraw("–ö–∞—Ç–µ—Ä–∏–Ω–∞ Sav—á—É–∫", -20)


def test_withdraw_records_transaction():
    """–ó–Ω—è—Ç—Ç—è –∑–∞–ø–∏—Å—É—î—Ç—å—Å—è –≤ —ñ—Å—Ç–æ—Ä—ñ—é."""
    create_account("–†–æ–º–∞–Ω –ì–æ–Ω—á–∞—Ä", initial_balance=500)
    
    withdraw("–†–æ–º–∞–Ω –ì–æ–Ω—á–∞—Ä", 100)
    withdraw("–†–æ–º–∞–Ω –ì–æ–Ω—á–∞—Ä", 50)
    
    history = get_transaction_history("–†–æ–º–∞–Ω –ì–æ–Ω—á–∞—Ä")
    assert len(history) == 2
    assert "–ó–Ω—è—Ç—Ç—è: -100" in history[0]
    assert "–ó–Ω—è—Ç—Ç—è: -50" in history[1]
```

```python
# ============================================
# –¢–µ—Å—Ç–∏ –ø–µ—Ä–µ–∫–∞–∑—É –∫–æ—à—Ç—ñ–≤
# ============================================

def test_transfer_between_accounts():
    """–ü–µ—Ä–µ–∫–∞–∑ –∫–æ—à—Ç—ñ–≤ –º—ñ–∂ —Ä–∞—Ö—É–Ω–∫–∞–º–∏."""
    create_account("–í—ñ–¥–ø—Ä–∞–≤–Ω–∏–∫", initial_balance=500)
    create_account("–û—Ç—Ä–∏–º—É–≤–∞—á", initial_balance=100)
    
    # –ü–µ—Ä–µ–∫–∞–∑—É—î–º–æ 200 –≤—ñ–¥ –í—ñ–¥–ø—Ä–∞–≤–Ω–∏–∫–∞ –¥–æ –û—Ç—Ä–∏–º—É–≤–∞—á–∞
    transfer("–í—ñ–¥–ø—Ä–∞–≤–Ω–∏–∫", "–û—Ç—Ä–∏–º—É–≤–∞—á", 200)
    
    assert get_balance("–í—ñ–¥–ø—Ä–∞–≤–Ω–∏–∫") == 300  # 500 - 200
    assert get_balance("–û—Ç—Ä–∏–º—É–≤–∞—á") == 300   # 100 + 200


def test_transfer_records_in_both_accounts():
    """–ü–µ—Ä–µ–∫–∞–∑ –∑–∞–ø–∏—Å—É—î—Ç—å—Å—è –≤ —ñ—Å—Ç–æ—Ä—ñ—é –æ–±–æ—Ö —Ä–∞—Ö—É–Ω–∫—ñ–≤."""
    create_account("–Ü–≤–∞–Ω", initial_balance=1000)
    create_account("–ú–∞—Ä—ñ—è", initial_balance=500)
    
    transfer("–Ü–≤–∞–Ω", "–ú–∞—Ä—ñ—è", 300)
    
    sender_history = get_transaction_history("–Ü–≤–∞–Ω")
    receiver_history = get_transaction_history("–ú–∞—Ä—ñ—è")
    
    assert any("–ü–µ—Ä–µ–∫–∞–∑ –ú–∞—Ä—ñ—è" in t for t in sender_history)
    assert any("–ü–µ—Ä–µ–∫–∞–∑ –≤—ñ–¥ –Ü–≤–∞–Ω" in t for t in receiver_history)


def test_transfer_insufficient_funds_raises_error():
    """–ü–µ—Ä–µ–∫–∞–∑ –±—ñ–ª—å—à–µ –±–∞–ª–∞–Ω—Å—É –º–∞—î –≤–∏–∫–∏–¥–∞—Ç–∏ –ø–æ–º–∏–ª–∫—É."""
    create_account("–ü–µ—Ç—Ä–æ", initial_balance=100)
    create_account("–û–ª—å–≥–∞", initial_balance=0)
    
    with pytest.raises(ValueError, match="–ù–µ–¥–æ—Å—Ç–∞—Ç–Ω—å–æ –∫–æ—à—Ç—ñ–≤ –Ω–∞ —Ä–∞—Ö—É–Ω–∫—É"):
        transfer("–ü–µ—Ç—Ä–æ", "–û–ª—å–≥–∞", 200)


def test_transfer_nonexistent_sender_raises_error():
    """–ü–µ—Ä–µ–∫–∞–∑ –∑ –Ω–µ—ñ—Å–Ω—É—é—á–æ–≥–æ —Ä–∞—Ö—É–Ω–∫—É –≤–∏–∫–∏–¥–∞—î –ø–æ–º–∏–ª–∫—É."""
    create_account("–û—Ç—Ä–∏–º—É–≤–∞—á", initial_balance=100)
    
    with pytest.raises(ValueError, match="–≤—ñ–¥–ø—Ä–∞–≤–Ω–∏–∫–∞.*–Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ"):
        transfer("–ù–µ—ñ—Å–Ω—É—é—á–∏–π", "–û—Ç—Ä–∏–º—É–≤–∞—á", 50)


def test_transfer_nonexistent_receiver_raises_error():
    """–ü–µ—Ä–µ–∫–∞–∑ –Ω–∞ –Ω–µ—ñ—Å–Ω—É—é—á–∏–π —Ä–∞—Ö—É–Ω–æ–∫ –≤–∏–∫–∏–¥–∞—î –ø–æ–º–∏–ª–∫—É."""
    create_account("–í—ñ–¥–ø—Ä–∞–≤–Ω–∏–∫", initial_balance=100)
    
    with pytest.raises(ValueError, match="–æ—Ç—Ä–∏–º—É–≤–∞—á–∞.*–Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ"):
        transfer("–í—ñ–¥–ø—Ä–∞–≤–Ω–∏–∫", "–ù–µ—ñ—Å–Ω—É—é—á–∏–π", 50)


# ============================================
# –¢–µ—Å—Ç–∏ —ñ—Å—Ç–æ—Ä—ñ—ó —Ç—Ä–∞–Ω–∑–∞–∫—Ü—ñ–π
# ============================================

def test_get_transaction_history_empty():
    """–Ü—Å—Ç–æ—Ä—ñ—è –Ω–æ–≤–æ–≥–æ —Ä–∞—Ö—É–Ω–∫—É –ø–æ—Ä–æ–∂–Ω—è."""
    create_account("–ê–Ω–Ω–∞ –õ–µ–±—ñ–¥—å")
    
    history = get_transaction_history("–ê–Ω–Ω–∞ –õ–µ–±—ñ–¥—å")
    assert history == []


def test_get_transaction_history_multiple_operations():
    """–Ü—Å—Ç–æ—Ä—ñ—è –º—ñ—Å—Ç–∏—Ç—å –≤—Å—ñ –æ–ø–µ—Ä–∞—Ü—ñ—ó."""
    create_account("–ú–∏—Ö–∞–π–ª–æ –ö—Ä–∞–≤–µ—Ü—å", initial_balance=1000)
    
    deposit("–ú–∏—Ö–∞–π–ª–æ –ö—Ä–∞–≤–µ—Ü—å", 500)
    withdraw("–ú–∏—Ö–∞–π–ª–æ –ö—Ä–∞–≤–µ—Ü—å", 200)
    
    history = get_transaction_history("–ú–∏—Ö–∞–π–ª–æ –ö—Ä–∞–≤–µ—Ü—å")
    assert len(history) == 2
    assert "–ü–æ–ø–æ–≤–Ω–µ–Ω–Ω—è: +500" in history[0]
    assert "–ó–Ω—è—Ç—Ç—è: -200" in history[1]


# ============================================
# –¢–µ—Å—Ç–∏ –∑–∞–∫—Ä–∏—Ç—Ç—è —Ä–∞—Ö—É–Ω–∫—É
# ============================================

def test_close_account_returns_balance():
    """–ó–∞–∫—Ä–∏—Ç—Ç—è —Ä–∞—Ö—É–Ω–∫—É –ø–æ–≤–µ—Ä—Ç–∞—î –∑–∞–ª–∏—à–æ–∫ –∫–æ—à—Ç—ñ–≤."""
    create_account("–í—ñ–∫—Ç–æ—Ä –ó–∞–π—Ü–µ–≤", initial_balance=250)
    
    final_balance = close_account("–í—ñ–∫—Ç–æ—Ä –ó–∞–π—Ü–µ–≤")
    
    assert final_balance == 250


def test_close_account_removes_from_system():
    """–ü—ñ—Å–ª—è –∑–∞–∫—Ä–∏—Ç—Ç—è —Ä–∞—Ö—É–Ω–∫—É –π–æ–≥–æ –Ω–µ –º–æ–∂–Ω–∞ –∑–Ω–∞–π—Ç–∏."""
    create_account("–Ü—Ä–∏–Ω–∞ –ë—ñ–ª–æ—É—Å", initial_balance=100)
    close_account("–Ü—Ä–∏–Ω–∞ –ë—ñ–ª–æ—É—Å")
    
    # –°–ø—Ä–æ–±–∞ –æ—Ç—Ä–∏–º–∞—Ç–∏ –±–∞–ª–∞–Ω—Å –º–∞—î –≤–∏–∫–∏–Ω—É—Ç–∏ –ø–æ–º–∏–ª–∫—É
    with pytest.raises(ValueError, match="–Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ"):
        get_balance("–Ü—Ä–∏–Ω–∞ –ë—ñ–ª–æ—É—Å")


# ============================================
# –ö–æ–º–ø–ª–µ–∫—Å–Ω—ñ —Ç–µ—Å—Ç–∏
# ============================================

def test_complex_transaction_sequence():
    """–°–∫–ª–∞–¥–Ω–∞ –ø–æ—Å–ª—ñ–¥–æ–≤–Ω—ñ—Å—Ç—å –æ–ø–µ—Ä–∞—Ü—ñ–π."""
    create_account("–ú–∏–∫–æ–ª–∞ –ö—Ä–∞–≤—á–µ–Ω–∫–æ", initial_balance=1000)
    
    # –ü–æ–ø–æ–≤–Ω–µ–Ω–Ω—è
    deposit("–ú–∏–∫–æ–ª–∞ –ö—Ä–∞–≤—á–µ–Ω–∫–æ", 500)
    assert get_balance("–ú–∏–∫–æ–ª–∞ –ö—Ä–∞–≤—á–µ–Ω–∫–æ") == 1500
    
    # –ó–Ω—è—Ç—Ç—è
    withdraw("–ú–∏–∫–æ–ª–∞ –ö—Ä–∞–≤—á–µ–Ω–∫–æ", 300)
    assert get_balance("–ú–∏–∫–æ–ª–∞ –ö—Ä–∞–≤—á–µ–Ω–∫–æ") == 1200
    
    # –©–µ –ø–æ–ø–æ–≤–Ω–µ–Ω–Ω—è
    deposit("–ú–∏–∫–æ–ª–∞ –ö—Ä–∞–≤—á–µ–Ω–∫–æ", 800)
    assert get_balance("–ú–∏–∫–æ–ª–∞ –ö—Ä–∞–≤—á–µ–Ω–∫–æ") == 2000
    
    # –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ —ñ—Å—Ç–æ—Ä—ñ—ó (–º–∞—î –±—É—Ç–∏ 3 —Ç—Ä–∞–Ω–∑–∞–∫—Ü—ñ—ó)
    history = get_transaction_history("–ú–∏–∫–æ–ª–∞ –ö—Ä–∞–≤—á–µ–Ω–∫–æ")
    assert len(history) == 3


def test_multiple_accounts_transfers():
    """–ö—ñ–ª—å–∫–∞ –ø–µ—Ä–µ–∫–∞–∑—ñ–≤ –º—ñ–∂ —Ä—ñ–∑–Ω–∏–º–∏ —Ä–∞—Ö—É–Ω–∫–∞–º–∏."""
    create_account("–ê", initial_balance=1000)
    create_account("–ë", initial_balance=500)
    create_account("–í", initial_balance=200)
    
    # –ê ‚Üí –ë: 300
    transfer("–ê", "–ë", 300)
    assert get_balance("–ê") == 700
    assert get_balance("–ë") == 800
    
    # –ë ‚Üí –í: 400
    transfer("–ë", "–í", 400)
    assert get_balance("–ë") == 400
    assert get_balance("–í") == 600
    
    # –í ‚Üí –ê: 100
    transfer("–í", "–ê", 100)
    assert get_balance("–í") == 500
    assert get_balance("–ê") == 800


def test_realistic_banking_scenario():
    """–†–µ–∞–ª—ñ—Å—Ç–∏—á–Ω–∏–π —Å—Ü–µ–Ω–∞—Ä—ñ–π –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –±–∞–Ω–∫—ñ–≤—Å—å–∫–∏—Ö –æ–ø–µ—Ä–∞—Ü—ñ–π."""
    # –°—Ç–≤–æ—Ä—é—î–º–æ —Ç—Ä–∏ —Ä–∞—Ö—É–Ω–∫–∏
    create_account("–ë–æ—Ä–∏—Å", initial_balance=5000)
    create_account("–°–≤—ñ—Ç–ª–∞–Ω–∞", initial_balance=3000)
    create_account("–î–µ–Ω–∏—Å", initial_balance=1000)
    
    # –ë–æ—Ä–∏—Å –∑–Ω—ñ–º–∞—î –≥–æ—Ç—ñ–≤–∫—É
    withdraw("–ë–æ—Ä–∏—Å", 1000)
    assert get_balance("–ë–æ—Ä–∏—Å") == 4000
    
    # –°–≤—ñ—Ç–ª–∞–Ω–∞ –ø–æ–ø–æ–≤–Ω—é—î —Ä–∞—Ö—É–Ω–æ–∫
    deposit("–°–≤—ñ—Ç–ª–∞–Ω–∞", 2000)
    assert get_balance("–°–≤—ñ—Ç–ª–∞–Ω–∞") == 5000
    
    # –ë–æ—Ä–∏—Å –ø–µ—Ä–µ–∫–∞–∑—É—î –≥—Ä–æ—à—ñ –î–µ–Ω–∏—Å—É
    transfer("–ë–æ—Ä–∏—Å", "–î–µ–Ω–∏—Å", 500)
    assert get_balance("–ë–æ—Ä–∏—Å") == 3500
    assert get_balance("–î–µ–Ω–∏—Å") == 1500
    
    # –î–µ–Ω–∏—Å –ø–µ—Ä–µ–∫–∞–∑—É—î –°–≤—ñ—Ç–ª–∞–Ω—ñ
    transfer("–î–µ–Ω–∏—Å", "–°–≤—ñ—Ç–ª–∞–Ω–∞", 700)
    assert get_balance("–î–µ–Ω–∏—Å") == 800
    assert get_balance("–°–≤—ñ—Ç–ª–∞–Ω–∞") == 5700
    
    # –ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ, —â–æ –≤—Å—ñ –º–∞—é—Ç—å —ñ—Å—Ç–æ—Ä—ñ—é
    assert len(get_transaction_history("–ë–æ—Ä–∏—Å")) == 2    # –∑–Ω—è—Ç—Ç—è + –ø–µ—Ä–µ–∫–∞–∑
    assert len(get_transaction_history("–°–≤—ñ—Ç–ª–∞–Ω–∞")) == 2 # –ø–æ–ø–æ–≤–Ω–µ–Ω–Ω—è + –æ—Ç—Ä–∏–º–∞–Ω–Ω—è
    assert len(get_transaction_history("–î–µ–Ω–∏—Å")) == 2    # –æ—Ç—Ä–∏–º–∞–Ω–Ω—è + –≤—ñ–¥–ø—Ä–∞–≤–∫–∞
```
