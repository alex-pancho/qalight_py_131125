–¶—ñ —Å–∏–º–≤–æ–ª–∏ –ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –¥–æ–∑–≤–æ–ª—è—é—Ç—å –≤–∞–º —Å—Ç–≤–æ—Ä—é–≤–∞—Ç–∏ —Ä—ñ–∑–Ω–æ–º–∞–Ω—ñ—Ç–Ω—ñ —Ñ–æ—Ä–º–∞—Ç–∏ –¥–ª—è –≤—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è –¥–∞—Ç–∏ —Ç–∞ —á–∞—Å—É —É –≤–∞—à–æ–º—É –ø—Ä–æ–≥—Ä–∞–º–Ω–æ–º—É –∫–æ–¥—ñ –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º —Ñ—É–Ω–∫—Ü—ñ—ó `strftime()`. –û–±–∏—Ä–∞–π—Ç–µ —Ç—ñ, —è–∫—ñ –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—é—Ç—å –≤–∞—à–∏–º –ø–æ—Ç—Ä–µ–±–∞–º —É —Ñ–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—ñ –¥–∞–Ω–∏—Ö.

## –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –∑—ñ —Å—Ç—Ä–æ–∫–∏ —É —á–∞—Å strptime()

–Ø–∫—â–æ —á–∞—Å –ø—Ä–∏—Ö–æ–¥–∏—Ç—å —É —Ñ–æ—Ä–º–∞—Ç—ñ —Å—Ç—Ä–æ–∫–∏, —Ç–æ –≤–∏ –º–æ–∂–µ—Ç–µ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—é `strptime()` –∑ –º–æ–¥—É–ª—è `time`, —è–∫–∞ –ø—Ä–∞—Ü—é—î –∑—ñ —Å—Ç—Ä—É–∫—Ç—É—Ä–æ—é —á–∞—Å—É `struct_time`.

–û—Å—å –ø—Ä–∏–∫–ª–∞–¥–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó `strptime()` –∑ –º–æ–¥—É–ª—è `time` –¥–ª—è –ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ä—è–¥–∫–∞ —É —Å—Ç—Ä—É–∫—Ç—É—Ä—É —á–∞—Å—É:

```python
import time

# –†—è–¥–æ–∫ –∑ —á–∞—Å–æ–º
time_string = '2023-12-31 23:59:59'

# –§–æ—Ä–º–∞—Ç —Ä—è–¥–∫–∞
format_string = '%Y-%m-%d %H:%M:%S'

# –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ä—è–¥–∫–∞ —É —Å—Ç—Ä—É–∫—Ç—É—Ä—É —á–∞—Å—É
time_obj = time.strptime(time_string, format_string)

print(time_obj)
# –í–∏–≤—ñ–¥: time.struct_time(tm_year=2023, tm_mon=12, tm_mday=31, tm_hour=23, tm_min=59, tm_sec=59, tm_wday=6, tm_yday=365, tm_isdst=-1)
```

–£ —Ü—å–æ–º—É –ø—Ä–∏–∫–ª–∞–¥—ñ —Ä—è–¥–æ–∫ `'2023-12-31 23:59:59'` –ø–µ—Ä–µ—Ç–≤–æ—Ä—é—î—Ç—å—Å—è —É —Ñ–æ—Ä–º–∞—Ç—ñ `%Y-%m-%d %H:%M:%S`, –¥–µ `%Y` - —Ä—ñ–∫, `%m` - –º—ñ—Å—è—Ü—å, `%d` - –¥–µ–Ω—å, `%H` - –≥–æ–¥–∏–Ω–∞, `%M` - —Ö–≤–∏–ª–∏–Ω–∞, `%S` - —Å–µ–∫—É–Ω–¥–∞. –ü—ñ—Å–ª—è –≤–∏–∫–ª–∏–∫—É `strptime()` –æ—Ç—Ä–∏–º—É—î–º–æ –æ–±'—î–∫—Ç `struct_time`, —è–∫–∏–π –º–æ–∂–Ω–∞ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ –¥–ª—è –ø–æ–¥–∞–ª—å—à–æ—ó –æ–±—Ä–æ–±–∫–∏ —Ç–∞ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –¥–∞–Ω–∏—Ö —á–∞—Å—É.

### –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è time() —É datetime()

–î–ª—è –ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –æ–±'—î–∫—Ç—É `time` —É –æ–±'—î–∫—Ç `datetime` –≤ Python –º–æ–∂–Ω–∞ —Å–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏—Å—è —Ñ—É–Ω–∫—Ü—ñ—î—é `fromtimestamp()` –∫–ª–∞—Å—É `datetime`.  –î–∞–≤–∞–π—Ç–µ —Ä–æ–∑–≥–ª—è–Ω–µ–º–æ –ø—Ä–∏–∫–ª–∞–¥:

```python
import time
from datetime import datetime

# –°—Ç–≤–æ—Ä—é—î–º–æ –æ–±'—î–∫—Ç time –∑ –ø–æ—Ç–æ—á–Ω–∏–º —á–∞—Å–æ–º —É —Ñ–æ—Ä–º–∞—Ç—ñ timestamp()
now = time.time()
print(now)

# –ü–µ—Ä–µ—Ç–≤–æ—é—î–º–æ –æ–±—î–∫—Ç struct_time —É datetime 
my_datetime = datetime.fromtimestamp(now)
print("–ü–æ—Ç–æ—á–Ω–∞ –¥–∞—Ç–∞ —Ç–∞ —á–∞—Å:", my_datetime)
```

# **–ú–æ–¥—É–ª—å datetime**

–ú–æ–¥—É–ª—å `datetime` —É Python –Ω–∞–¥–∞—î –∑—Ä—É—á–Ω—ñ —ñ–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç–∏ –¥–ª—è –º–æ–¥–µ–ª—é–≤–∞–Ω–Ω—è –æ–±'—î–∫—Ç—ñ–≤ –¥–∞—Ç–∏ —Ç–∞ —á–∞—Å—É, –≤–∫–ª—é—á–∞—é—á–∏ –ø—ñ–¥—Ç—Ä–∏–º–∫—É —á–∞—Å–æ–≤–∏—Ö –ø–æ—è—Å—ñ–≤. –û—Å–Ω–æ–≤–Ω—ñ –∫–ª–∞—Å–∏ —ñ –º–µ—Ç–æ–¥–∏ —Ü—å–æ–≥–æ –º–æ–¥—É–ª—è –¥–æ–∑–≤–æ–ª—è—é—Ç—å —Å—Ç–≤–æ—Ä—é–≤–∞—Ç–∏, —Ñ–æ—Ä–º–∞—Ç—É–≤–∞—Ç–∏ —Ç–∞ –æ–ø–µ—Ä—É–≤–∞—Ç–∏ –æ–±'—î–∫—Ç–∞–º–∏ –¥–∞—Ç–∏ —Ç–∞ —á–∞—Å—É –∑ —É—Ä–∞—Ö—É–≤–∞–Ω–Ω—è–º —á–∞—Å–æ–≤–∏—Ö –ø–æ—è—Å—ñ–≤.

2. **–ö–ª–∞—Å `datetime.timezone`**: –¶–µ–π –∫–ª–∞—Å —î –ø—Ä–æ—Å—Ç–æ—é —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—î—é —á–∞—Å–æ–≤–æ–≥–æ –ø–æ—è—Å—É —É –º–æ–¥—É–ª—ñ `datetime`. –í—ñ–Ω –¥–æ–∑–≤–æ–ª—è—î —Å—Ç–≤–æ—Ä—é–≤–∞—Ç–∏ –æ–±'—î–∫—Ç–∏ —á–∞—Å—É –∑ —É—Ä–∞—Ö—É–≤–∞–Ω–Ω—è–º —á–∞—Å–æ–≤–æ–≥–æ –ø–æ—è—Å—É.

–ú–æ–¥—É–ª—å `zoneinfo`, —è–∫–∏–π –∑'—è–≤–∏–≤—Å—è —É –Ω–æ–≤—ñ—à–∏—Ö –≤–µ—Ä—Å—ñ—è—Ö Python, –ø—Ä–æ–ø–æ–Ω—É—î –±—ñ–ª—å—à –±–∞–≥–∞—Ç—É –∫–æ–Ω–∫—Ä–µ—Ç–Ω—É —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—é `tzinfo`, —â–æ –¥–æ–∑–≤–æ–ª—è—î –ø—Ä–∞—Ü—é–≤–∞—Ç–∏ –∑ –±—ñ–ª—å—à–∏–º —Ä–æ–∑–º–∞—ó—Ç—Ç—è–º —á–∞—Å–æ–≤–∏—Ö –ø–æ—è—Å—ñ–≤ —Ç–∞ –ø—Ä–∞–≤–∏–ª –ø–µ—Ä–µ—Ö–æ–¥—É –Ω–∞ –ª—ñ—Ç–Ω—ñ–π/–∑–∏–º–æ–≤–∏–π —á–∞—Å. –¶–µ –æ—Å–æ–±–ª–∏–≤–æ –∫–æ—Ä–∏—Å–Ω–æ –ø—Ä–∏ —Ä–æ–±–æ—Ç—ñ –∑ –¥–∞–Ω–∏–º–∏, –¥–µ –≤–∞–∂–ª–∏–≤–∞ —Ç–æ—á–Ω–∞ —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—è –ø—Ä–æ —á–∞—Å —ñ —á–∞—Å–æ–≤—ñ –ø–æ—è—Å–∏.

## –ú–µ—Ç–æ–¥–∏ –º–æ–¥—É–ª—è **datetime**

### –ú–µ—Ç–æ–¥ `fromordinal()`

–ú–µ—Ç–æ–¥ `fromordinal()` —É –∫–ª–∞—Å—ñ `datetime` –º–æ–¥–µ–ª—é—î –æ–±'—î–∫—Ç –¥–∞—Ç–∏ –∑ —ó—ó –ø–æ—Ä—è–¥–∫–æ–≤–∏–º –Ω–æ–º–µ—Ä–æ–º, –¥–µ –¥–Ω—ñ –≤—ñ–¥–ª—ñ—á—É—é—Ç—å—Å—è –≤—ñ–¥ 1 —Å—ñ—á–Ω—è 1 —Ä–æ–∫—É. –¶–µ–π –º–µ—Ç–æ–¥ –¥–æ–∑–≤–æ–ª—è—î —Å—Ç–≤–æ—Ä—é–≤–∞—Ç–∏ –æ–±'—î–∫—Ç–∏ `datetime` –Ω–∞ –æ—Å–Ω–æ–≤—ñ –ø–æ—Ä—è–¥–∫–æ–≤–æ–≥–æ –Ω–æ–º–µ—Ä–∞ –¥–Ω—è —É —Ä–æ—Ü—ñ.

–û—Å—å –ø—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –º–µ—Ç–æ–¥—É `fromordinal()`:

```python
from datetime import datetime

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –æ–±'—î–∫—Ç–∞ datetime –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –ø–æ—Ä—è–¥–∫–æ–≤–æ–≥–æ –Ω–æ–º–µ—Ä–∞ –¥–Ω—è
ordinal_day = 365  # –ü–æ—Ä—è–¥–∫–æ–≤–∏–π –Ω–æ–º–µ—Ä –¥–Ω—è
dt = datetime.fromordinal(ordinal_day)

print(dt)
# –í–∏–≤—ñ–¥: 0001-12-31 00:00:00 # 31 –≥—Ä—É–¥–Ω—è 1 —Ä–æ–∫—É

```

–£ —Ü—å–æ–º—É –ø—Ä–∏–∫–ª–∞–¥—ñ –º–µ—Ç–æ–¥ `fromordinal()` —Å—Ç–≤–æ—Ä—é—î –æ–±'—î–∫—Ç `datetime`, —è–∫–∏–π –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—î 365-–π –¥–µ–Ω—å (–æ—Å—Ç–∞–Ω–Ω—ñ–π –¥–µ–Ω—å) —É —Ä–æ—Ü—ñ. –ë—É–¥—å-—è–∫–∏–π –¥–µ–Ω—å —É —Ä–æ—Ü—ñ –º–æ–∂–µ –±—É—Ç–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–π —Å–≤–æ—ó–º –ø–æ—Ä—è–¥–∫–æ–≤–∏–º –Ω–æ–º–µ—Ä–æ–º, —ñ —Ü–µ–π –º–µ—Ç–æ–¥ –¥–æ–∑–≤–æ–ª—è—î –ª–µ–≥–∫–æ —Å—Ç–≤–æ—Ä—é–≤–∞—Ç–∏ –æ–±'—î–∫—Ç–∏ `datetime` –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º —Ü—å–æ–≥–æ –Ω–æ–º–µ—Ä–∞.

### –ó–≤–æ—Ä–æ–Ω—ñ–π –∑–∞ –¥—ñ—î—é –º–µ—Ç–æ–¥ - `.toordinal()`

–ú–µ—Ç–æ–¥ `.toordinal()`, —è–∫–∏–π –¥–æ–∑–≤–æ–ª—è—î –æ—Ç—Ä–∏–º–∞—Ç–∏ –ø–æ—Ä—è–¥–∫–æ–≤–∏–π –Ω–æ–º–µ—Ä –¥–Ω—è —É —Ä–æ—Ü—ñ –±–µ–∑–ø–æ—Å–µ—Ä–µ–¥–Ω—å–æ –∑ –æ–±'—î–∫—Ç–∞ –¥–∞—Ç–∏. –í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è —Ü—å–æ–≥–æ –º–µ—Ç–æ–¥—É —î –¥–æ—Å–∏—Ç—å –ø—Ä–æ—Å—Ç–∏–º —ñ –µ—Ñ–µ–∫—Ç–∏–≤–Ω–∏–º –¥–ª—è –æ—Ç—Ä–∏–º–∞–Ω–Ω—è –ø–æ—Ä—è–¥–∫–æ–≤–æ–≥–æ –Ω–æ–º–µ—Ä–∞ –¥–Ω—è —É —Ä–æ—Ü—ñ.

–û—Å—å –ø—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –º–µ—Ç–æ–¥—É `.toordinal()` –¥–ª—è –æ—Ç—Ä–∏–º–∞–Ω–Ω—è –ø–æ—Ä—è–¥–∫–æ–≤–æ–≥–æ –Ω–æ–º–µ—Ä–∞ –¥–Ω—è —É —Ä–æ—Ü—ñ:

```python
from datetime import datetime

day_number = datetime.today().toordinal()
print(f'–ü–æ—Ä—è–¥–∫–æ–≤–∏–π –Ω–æ–º–µ—Ä —Å—å–æ–≥–æ–¥–Ω—ñ—à–Ω—å–æ–≥–æ –¥–Ω—è: {day_number}')
```

–£ —Ü—å–æ–º—É –ø—Ä–∏–∫–ª–∞–¥—ñ –º–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î–º–æ –º–µ—Ç–æ–¥ `toordinal()` –¥–ª—è –æ—Ç—Ä–∏–º–∞–Ω–Ω—è –ø–æ—Ä—è–¥–∫–æ–≤–æ–≥–æ –Ω–æ–º–µ—Ä–∞ —Å—å–æ–≥–æ–¥–Ω—ñ—à–Ω–æ–≥–æ –¥–Ω—è –≤—ñ–¥ 1 —Å—ñ—á–Ω—è 1 —Ä–æ–∫—É.

```python
from datetime import datetime

def day_of_year(date_str):
    # –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ä—è–¥–∫–∞ —É –æ–±'—î–∫—Ç datetime
    date_obj = datetime.strptime(date_str, '%Y-%m-%d').toordinal()
    
    # –û—Ç—Ä–∏–º–∞–Ω–Ω—è –ø–æ—Ä—è–¥–∫–æ–≤–æ–≥–æ –Ω–æ–º–µ—Ä–∞ –¥–Ω—è
    day_number = date_obj - datetime(date_obj.year, 1, 1).toordinal() + 1
    
    return day_number

# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–ª–∏–∫—É —Ñ—É–Ω–∫—Ü—ñ—ó
date_str = '2023-12-31'
day_number = day_of_year(date_str)
print(f'–ü–æ—Ä—è–¥–∫–æ–≤–∏–π –Ω–æ–º–µ—Ä –¥–Ω—è –¥–ª—è {date_str}: {day_number}')
```

–£ —Ü—å–æ–º—É –ø—Ä–∏–∫–ª–∞–¥—ñ:

- –ú–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î–º–æ –º–µ—Ç–æ–¥ `toordinal()` –¥–ª—è –æ—Ç—Ä–∏–º–∞–Ω–Ω—è –ø–æ—Ä—è–¥–∫–æ–≤–æ–≥–æ –Ω–æ–º–µ—Ä–∞ –¥–Ω—è —É —Ä–æ—Ü—ñ.
- –í–∏—Ä–∞—Ö–æ–≤—É—î–º–æ —Ä—ñ–∑–Ω–∏—Ü—é –º—ñ–∂ –ø–æ—Ä—è–¥–∫–æ–≤–∏–º–∏ –Ω–æ–º–µ—Ä–∞–º–∏ –ø–æ—Ç–æ—á–Ω–æ—ó –¥–∞—Ç–∏ —Ç–∞ –ø–µ—Ä—à–æ–≥–æ –¥–Ω—è —Ä–æ–∫—É, —â–æ–± –æ—Ç—Ä–∏–º–∞—Ç–∏ –∫–æ–Ω–∫—Ä–µ—Ç–Ω–∏–π –ø–æ—Ä—è–¥–∫–æ–≤–∏–π –Ω–æ–º–µ—Ä –¥–Ω—è —É —Ä–æ—Ü—ñ.

–í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –º–µ—Ç–æ–¥—É `toordinal()` –º–æ–∂–µ –±—É—Ç–∏ –∑—Ä—É—á–Ω–∏–º, –æ—Å–æ–±–ª–∏–≤–æ —è–∫—â–æ –≤–∞–º –ø–æ—Ç—Ä—ñ–±–Ω–æ –ø—Ä–∞—Ü—é–≤–∞—Ç–∏ –≤–∏–∫–ª—é—á–Ω–æ –∑ –ø–æ—Ä—è–¥–∫–æ–≤–∏–º –Ω–æ–º–µ—Ä–æ–º –¥–Ω—è —É —Ä–æ—Ü—ñ.

### –ú–µ—Ç–æ–¥ `fromtimestamp()`

–ú–µ—Ç–æ–¥ `fromtimestamp()` –≤ –º–æ–¥—É–ª—ñ `datetime` —É Python –¥–æ–∑–≤–æ–ª—è—î —Å—Ç–≤–æ—Ä—é–≤–∞—Ç–∏ –æ–±'—î–∫—Ç –¥–∞—Ç–∏ —Ç–∞ —á–∞—Å—É –∑ –∑–∞–¥–∞–Ω–æ–≥–æ –º—ñ—Ç–∫–∏ —á–∞—Å—É (timestamp). –ú—ñ—Ç–∫–∞ —á–∞—Å—É –≤–∏–∑–Ω–∞—á–∞—î –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Å–µ–∫—É–Ω–¥, —â–æ –º–∏–Ω—É–ª–∏ –∑ –ø–µ–≤–Ω–æ–≥–æ –ø–æ—á–∞—Ç–∫–æ–≤–æ–≥–æ –º–æ–º–µ–Ω—Ç—É (–∑–∞–∑–≤–∏—á–∞–π –∑—ñ —Å—Ç–∞—Ä—Ç—É UNIX Epoch - 1 —Å—ñ—á–Ω—è 1970 —Ä–æ–∫—É 00:00:00 UTC).

–û—Å—å –ø—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –º–µ—Ç–æ–¥—É `fromtimestamp()` –¥–ª—è —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –æ–±'—î–∫—Ç–∞ –¥–∞—Ç–∏ —Ç–∞ —á–∞—Å—É –∑ –º—ñ—Ç–∫–∏ —á–∞—Å—É:

```python
from datetime import datetime

# –ü—Ä–∏–∫–ª–∞–¥ –º—ñ—Ç–∫–∏ —á–∞—Å—É (timestamp)
timestamp = 1634521800  # UNIX Epoch: 18 –∂–æ–≤—Ç–Ω—è 2021 —Ä–æ–∫—É 10:30:00 UTC

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –æ–±'—î–∫—Ç–∞ datetime –∑ –º—ñ—Ç–∫–∏ —á–∞—Å—É
dt = datetime.fromtimestamp(timestamp)

print(dt)
# –í–∏–≤–µ–¥–µ: 2021-10-18 10:30:00

```

–£ —Ü—å–æ–º—É –ø—Ä–∏–∫–ª–∞–¥—ñ –º–µ—Ç–æ–¥ `fromtimestamp()` —Å—Ç–≤–æ—Ä—é—î –æ–±'—î–∫—Ç `datetime`, –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –º—ñ—Ç–∫—É —á–∞—Å—É `timestamp`. –í–∞–∂–ª–∏–≤–æ –≤—Ä–∞—Ö–æ–≤—É–≤–∞—Ç–∏, —â–æ –º—ñ—Ç–∫–∞ —á–∞—Å—É –º–æ–∂–µ –±—É—Ç–∏ –≤—ñ–¥–Ω–æ—Å–Ω–æ—é (–≤—ñ–¥–ª—ñ–∫ –≤—ñ–¥ –ø–æ—á–∞—Ç–∫–æ–≤–æ–≥–æ –º–æ–º–µ–Ω—Ç—É) –∞–±–æ –∞–±—Å–æ–ª—é—Ç–Ω–æ—é (–∞–±—Å–æ–ª—é—Ç–Ω–∏–π —á–∞—Å –≤—ñ–¥–Ω–æ—Å–Ω–æ UNIX Epoch, –Ω–∞–ø—Ä–∏–∫–ª–∞–¥).

### **–ú–µ—Ç–æ–¥ today()**

–ú–µ—Ç–æ–¥ `today()` —É –º–æ–¥—É–ª—ñ `datetime` —É Python –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –æ—Ç—Ä–∏–º–∞–Ω–Ω—è –ø–æ—Ç–æ—á–Ω–æ—ó –¥–∞—Ç–∏ (–±–µ–∑ —á–∞—Å—É) –∑–≥—ñ–¥–Ω–æ –∑ –ª–æ–∫–∞–ª—å–Ω–∏–º —á–∞—Å–æ–≤–∏–º –ø–æ—è—Å–æ–º –≤–∞—à–æ—ó —Å–∏—Å—Ç–µ–º–∏. –í—ñ–Ω –ø–æ–≤–µ—Ä—Ç–∞—î –æ–±'—î–∫—Ç `datetime.date`, —è–∫–∏–π –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—î —Å—å–æ–≥–æ–¥–Ω—ñ—à–Ω—é –¥–∞—Ç—É –±–µ–∑ –≤—Ä–∞—Ö—É–≤–∞–Ω–Ω—è —á–∞—Å—É.

–û—Å—å –ø—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –º–µ—Ç–æ–¥—É `today()`:

```python
from datetime import date

# –û—Ç—Ä–∏–º–∞–Ω–Ω—è –ø–æ—Ç–æ—á–Ω–æ—ó –¥–∞—Ç–∏
today_date = date.today()
print(today_date)
# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–≤–µ–¥–µ–Ω–Ω—è: 2023-12-27
```

–£ –±—ñ–ª—å—à–æ—Å—Ç—ñ –≤–∏–ø–∞–¥–∫—ñ–≤, `timedelta` —Å—Ç–≤–æ—Ä—é—î—Ç—å—Å—è –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –≤—ñ–¥–Ω—ñ–º–∞–Ω–Ω—è –¥–≤–æ—Ö –æ–±'—î–∫—Ç—ñ–≤ –¥–∞—Ç–∏, —á–∞—Å—É –∞–±–æ `datetime`, —â–æ–± –≤–∏–∑–Ω–∞—á–∏—Ç–∏ —Ä—ñ–∑–Ω–∏—Ü—é –º—ñ–∂ –Ω–∏–º–∏. –û—Å—å –ø—Ä–∏–∫–ª–∞–¥:

```python
from datetime import datetime, timedelta

# –ü–æ—Ç–æ—á–Ω–∞ –¥–∞—Ç–∞ —Ç–∞ —á–∞—Å
now = datetime.now()
# –ü–æ–ø–µ—Ä–µ–¥–Ω—è –¥–∞—Ç–∞ —Ç–∞ —á–∞—Å (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, —è–∫—â–æ –≤–∏ —Ö–æ—á–µ—Ç–µ –≤–∏–∑–Ω–∞—á–∏—Ç–∏, —â–æ –±—É–ª–æ 5 –¥–Ω—ñ–≤ —Ç–æ–º—É)
past_date = now - timedelta(days=5)
# –†—ñ–∑–Ω–∏—Ü—è –º—ñ–∂ –ø–æ—Ç–æ—á–Ω–æ—é —Ç–∞ –ø–æ–ø–µ—Ä–µ–¥–Ω—å–æ—é –¥–∞—Ç–æ—é
time_difference = now - past_date

print(f"–ü–æ—Ç–æ—á–Ω–∞ –¥–∞—Ç–∞ —Ç–∞ —á–∞—Å: {now}")
print(f"–ü–æ–ø–µ—Ä–µ–¥–Ω—è –¥–∞—Ç–∞ —Ç–∞ —á–∞—Å: {past_date}")
print(f"–†—ñ–∑–Ω–∏—Ü—è –º—ñ–∂ –¥–∞—Ç–∞–º–∏: {time_difference}")
```

–£ —Ü—å–æ–º—É –ø—Ä–∏–∫–ª–∞–¥—ñ `timedelta` –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è —Ä—ñ–∑–Ω–∏—Ü—ñ –º—ñ–∂ –ø–æ—Ç–æ—á–Ω–æ—é –¥–∞—Ç–æ—é —Ç–∞ –¥–∞—Ç–æ—é, —è–∫–∞ –≤—ñ–¥–±—É–ª–∞—Å—è 5 –¥–Ω—ñ–≤ —Ç–æ–º—É. –†–µ–∑—É–ª—å—Ç–∞—Ç - —Ä—ñ–∑–Ω–∏—Ü—è –≤ —á–∞—Å—ñ –º—ñ–∂ —Ü–∏–º–∏ –¥–≤–æ–º–∞ –º–æ–º–µ–Ω—Ç–∞–º–∏. –¢–∞–∫—ñ –æ–ø–µ—Ä–∞—Ü—ñ—ó –∑—Ä—É—á–Ω–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ –¥–ª—è –æ–±—á–∏—Å–ª–µ–Ω–Ω—è —ñ–Ω—Ç–µ—Ä–≤–∞–ª—ñ–≤ —á–∞—Å—É —Ç–∞ –∫–µ—Ä—É–≤–∞–Ω–Ω—è —á–∞—Å–æ–≤–∏–º–∏ –¥–∞–Ω–∏–º–∏ –≤ –ø—Ä–æ–≥—Ä–∞–º–∞—Ö Python.

–î–ª—è —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—ó —É–º–æ–≤–∏, —â–æ–± –≤–∏–≤–æ–¥–∏—Ç–∏ WARNING, —è–∫—â–æ —Ä—ñ–∑–Ω–∏—Ü—è –º—ñ–∂ —á–∞—Å–æ–º A —ñ —á–∞—Å–æ–º –ë –±—ñ–ª—å—à–µ 31 —Å–µ–∫—É–Ω–¥–∏, –º–æ–∂–Ω–∞ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏ –º–æ–¥—É–ª—å `datetime` –¥–ª—è —Ä–æ–±–æ—Ç–∏ –∑ —á–∞—Å–æ–º —Ç–∞ `timedelta` –¥–ª—è –æ–±—á–∏—Å–ª–µ–Ω–Ω—è —Ä—ñ–∑–Ω–∏—Ü—ñ —É —á–∞—Å—ñ.k.

–û—Å—å –Ω–µ–≤–µ–ª–∏–∫–∏–π –ø—Ä–∏–∫–ª–∞–¥, —è–∫ —Ü–µ –º–æ–∂–Ω–∞ –∑—Ä–æ–±–∏—Ç–∏ –≤ Python:

```python
from datetime import datetime, timedelta

def is_diff_more_30_sec(time_A:datetime, time_B:datetime):
		# –û–±—á–∏—Å–ª–µ–Ω–Ω—è —Ä—ñ–∑–Ω–∏—Ü—ñ —É —á–∞—Å—ñ
      time_difference = time_B - time_A
		# –ø–æ–≤–µ—Ä–Ω–µ True, —è–∫—â–æ —Ä—ñ–∑–Ω–∏—Ü—è –±—ñ–ª—å—à–µ 31 —Å–µ–∫—É–Ω–¥–∏
      return time_difference > timedelta(seconds=31)

# –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ —Ä–æ–±–æ—Ç–∏
time_a = datetime.now()  # –æ—Ç—Ä–∏–º–∞—î–º–æ –ø–æ—Ç–æ—á–Ω–∏–π —á–∞—Å
time_b = time_a + timedelta(seconds=32) # —à—Ç—É—á–Ω–æ –∑–±—ñ–ª—å—à–∏–º–æ –π–æ–≥–æ –Ω–∞ 32 —Å–µ–∫—É–Ω–¥–∏
if is_diff_more_30_sec(time_a, time_b):
   print("WARNING! –†—ñ–∑–Ω–∏—Ü—è –±—ñ–ª—å—à–µ 31 —Å–µ–∫—É–Ω–¥!") 
```

- `time_A` –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—î —á–∞—Å –ø–æ—á–∞—Ç–∫—É.
- `time_B` –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—î —á–∞—Å –ø—ñ—Å–ª—è –∑–∞—Ç—Ä–∏–º–∫–∏.
- `time_difference` –æ–±—á–∏—Å–ª—é—î —Ä—ñ–∑–Ω–∏—Ü—é —É —á–∞—Å—ñ –º—ñ–∂ `time_A` —Ç–∞ `time_B`.
- –£–º–æ–≤–∞ –ø–µ—Ä–µ–≤—ñ—Ä—è—î, —á–∏ —Ä—ñ–∑–Ω–∏—Ü—è –±—ñ–ª—å—à–µ 31 —Å–µ–∫—É–Ω–¥–∏ —ñ –≤–∏–≤–æ–¥–∏—Ç—å "WARNING!" —É –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ–º—É –≤–∏–ø–∞–¥–∫—É.
