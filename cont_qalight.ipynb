# Програма курсу “QA Automation on Python”

# Блок1: Програмування на Python

1. Знайомство. Загальні відомості та все, щоб почати роботу
1. Робота з Git. Репозиторій Git Hub, як основа співпраці програмістів
1. Синтаксис, структура, ключові слова. Типи даних. Числові операції. Ввід/Вивід даних
1. Строки `str`: методи, атрибути та корисні трюки
1. Умови (`if/elif/else`) та цикли (`for/while`)
1. Колекції: `tuple`, `list`, `set`, `dict`
1. Створення і робота функцій (`def`)
1. Виключення та робота з ними(`try/except/finaly`). Використання `assert` у програмі
1. Робота з файлами, бібліотека `pathlib` та її модуль `Path`. Імпорт коду  (частина 1)
1. Практична обробка `csv, json, xml` файлів.


# Блок 2: Тестування та обєктно-орієнтований підхід.

1. Вступ у тестування на `pytest`. Імпорт коду (частина 2)
1. Вступ до ООП, Класи (`class`) та об'єкти, атрибути класу, наслідування.
1. Життєвий цикл класу. Магічні методи, інкапсуляція і поліморфізм. Області видимості змінних
1. Ітератори та генератори. Декоратори, створення декораторів.
1. Тестування REST API, частина 1: Робота з мережею. Бібліотеки `urllib` та `requests`. Запити `get, post, put, delete`
1. Тестування REST API, частина 2: Перевірка відповідності OASv3. End-2-End тестування. Структура тестувального фреймворку
1. SQL і Робота з реляційними базами даних (`sqlite3`, MySQL) Робота з базами даних через ORM (`SQLAlchemy`)
1. Дата і час (`time, datetime, timedelta`)
1. Використання патернів та декораторів(фікстур) у pytest. Параметризовані тести.
1. Парсинг даних у тестуванні. `XPath`, `CSS`-локатори. Регулярні вирази
1. Вступ у Selenium - робота з веб елементами. Налаштування роботи браузера. Віддалене виконання у Selenium Grid
1. Очікування на веб-елемент у WebDriver. Основи побудови PageObject
1. CI/CD на прикладі Jenkins (freestyle and pipeline)
1. Final countdown (лекція з настановами та рекомендаціями щодо пошуку роботи)
