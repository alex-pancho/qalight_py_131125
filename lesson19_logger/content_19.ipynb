```python
import logging

# –ù–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è –∫–æ–Ω—Ñ—ñ–≥—É—Ä–∞—Ü—ñ—ó –ª–æ–≥—É–≤–∞–Ω–Ω—è
filenamelogging.basicConfig(='example.log', level=logging.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s')

# –õ–æ–≥—É–≤–∞–Ω–Ω—è –ø–æ–¥—ñ–π —Ä—ñ–∑–Ω–æ–≥–æ —Ä—ñ–≤–Ω—è (DEBUG, INFO, WARNING, ERROR, CRITICAL)
logging.debug('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è DEBUG')
logging.info('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è INFO')
logging.warning('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è WARNING')
logging.error('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è ERROR')
logging.critical('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è CRITICAL')

```

–Ø–∫—â–æ –≤–∏ —Ö–æ—á–µ—Ç–µ –ª–æ–≥—É–≤–∞—Ç–∏ –ø–æ–¥—ñ—ó –≤ –∫–æ–Ω—Å–æ–ª—å, –º–æ–∂–µ—Ç–µ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ `StreamHandler`:

```python
import logging

# –ù–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è –∫–æ–Ω—Ñ—ñ–≥—É—Ä–∞—Ü—ñ—ó –ª–æ–≥—É–≤–∞–Ω–Ω—è
logging.basicConfig(level=logging.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s')

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –æ–±—Ä–æ–±–Ω–∏–∫–∞ –¥–ª—è –≤–∏–≤–æ–¥—É –≤ –∫–æ–Ω—Å–æ–ª—å
console_handler = logging.StreamHandler()
console_handler.setLevel(logging.DEBUG)
logging.getLogger('').addHandler(console_handler)

# –õ–æ–≥—É–≤–∞–Ω–Ω—è –ø–æ–¥—ñ–π —Ä—ñ–∑–Ω–æ–≥–æ —Ä—ñ–≤–Ω—è
logging.debug('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è DEBUG')
logging.info('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è INFO')
logging.warning('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è WARNING')
logging.error('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è ERROR')
logging.critical('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è CRITICAL')

```

## –ó–∞—Ö–≤–∞—Ç —É —Ñ–∞–π–ª

–©–æ–± —Å—Ç–≤–æ—Ä–∏—Ç–∏ –æ–±—Ä–æ–±–Ω–∏–∫ –¥–ª—è –∑–∞–ø–∏—Å—É –≤ —Ñ–∞–π–ª –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –º–æ–¥—É–ª—è `logging`, –≤–∞–º —Å–ª—ñ–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ –∫–ª–∞—Å `FileHandler`. –û—Å—å –ø—Ä–∏–∫–ª–∞–¥, —è–∫ –≤–∏ –º–æ–∂–µ—Ç–µ —Ü–µ –∑—Ä–æ–±–∏—Ç–∏:

```python
import logging

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –ª–æ–≥–µ—Ä–∞
logger = logging.getLogger(__name__)

# –ù–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è —Ä—ñ–≤–Ω—è –ª–æ–≥—É–≤–∞–Ω–Ω—è
logger.setLevel(logging.DEBUG)

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –æ–±—Ä–æ–±–Ω–∏–∫–∞ –¥–ª—è –∑–∞–ø–∏—Å—É –≤ —Ñ–∞–π–ª
file_handler = logging.FileHandler('logfile.txt')

# –ù–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è —Ä—ñ–≤–Ω—è –ª–æ–≥—É–≤–∞–Ω–Ω—è –¥–ª—è –æ–±—Ä–æ–±–Ω–∏–∫–∞
file_handler.setLevel(logging.DEBUG)

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ñ–æ—Ä–º–∞—Ç–µ—Ä–∞ –¥–ª—è –æ–±—Ä–æ–±–Ω–∏–∫–∞
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
file_handler.setFormatter(formatter)

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –æ–±—Ä–æ–±–Ω–∏–∫–∞ –¥–æ –ª–æ–≥–µ—Ä–∞
logger.addHandler(file_handler)

# –õ–æ–≥—É–≤–∞–Ω–Ω—è –ø–æ–¥—ñ–π
logger.debug('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è DEBUG')
logger.info('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è INFO')
logger.warning('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è WARNING')
logger.error('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è ERROR')
logger.critical('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è CRITICAL')

```

## –ó–∞—Ö–≤–∞—Ç stdout

–©–æ–± —Å—Ç–≤–æ—Ä–∏—Ç–∏ –æ–±—Ä–æ–±–Ω–∏–∫ –¥–ª—è –≤–∏–≤–æ–¥—É –≤ `stdout` (—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∏–π –≤–∏–≤—ñ–¥), –≤–∏ –º–æ–∂–µ—Ç–µ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ –∫–ª–∞—Å `StreamHandler` –∑ –º–æ–¥—É–ª—è `logging`. –û—Å—å –ø—Ä–∏–∫–ª–∞–¥:

```python
import logging

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –ª–æ–≥–µ—Ä–∞
logger = logging.getLogger(__name__)

# –ù–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è —Ä—ñ–≤–Ω—è –ª–æ–≥—É–≤–∞–Ω–Ω—è
logger.setLevel(logging.DEBUG)

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –æ–±—Ä–æ–±–Ω–∏–∫–∞ –¥–ª—è –≤–∏–≤–æ–¥—É –≤ stdout
console_handler = logging.StreamHandler()

# –ù–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è —Ä—ñ–≤–Ω—è –ª–æ–≥—É–≤–∞–Ω–Ω—è –¥–ª—è –æ–±—Ä–æ–±–Ω–∏–∫–∞
console_handler.setLevel(logging.DEBUG)

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ñ–æ—Ä–º–∞—Ç–µ—Ä–∞ –¥–ª—è –æ–±—Ä–æ–±–Ω–∏–∫–∞
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
console_handler.setFormatter(formatter)

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –æ–±—Ä–æ–±–Ω–∏–∫–∞ –¥–æ –ª–æ–≥–µ—Ä–∞
logger.addHandler(console_handler)

# –õ–æ–≥—É–≤–∞–Ω–Ω—è –ø–æ–¥—ñ–π
logger.debug('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è DEBUG')
logger.info('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è INFO')
logger.warning('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è WARNING')
logger.error('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è ERROR')
logger.critical('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è CRITICAL')

```

–£ —Ü—å–æ–º—É –ø—Ä–∏–∫–ª–∞–¥—ñ —Å—Ç–≤–æ—Ä—é—î—Ç—å—Å—è –æ–±—Ä–æ–±–Ω–∏–∫ `StreamHandler`, —è–∫–∏–π –≤–∏–≤–æ–¥–∏—Ç—å –ª–æ–≥–∏ –Ω–∞ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∏–π –≤–∏–≤—ñ–¥ (`stdout`). –í–∏ –º–æ–∂–µ—Ç–µ –Ω–∞–ª–∞—à—Ç—É–≤–∞—Ç–∏ —Ä—ñ–≤–µ–Ω—å –ª–æ–≥—É–≤–∞–Ω–Ω—è, –≤–∏–±—Ä–∞—Ç–∏ —Ñ–æ—Ä–º–∞—Ç –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω—å —Ç–∞ –¥–æ–¥–∞—Ç–∏ —Ü–µ–π –æ–±—Ä–æ–±–Ω–∏–∫ –¥–æ –ª–æ–≥–µ—Ä–∞.

–Ø–∫—â–æ –≤–∞–º –Ω–µ –ø–æ—Ç—Ä—ñ–±–Ω–æ –≤–∏–≤–æ–¥–∏—Ç–∏ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –∑ –ø–µ–≤–Ω–æ–≥–æ —Ä—ñ–≤–Ω—è, –≤–∏ –º–æ–∂–µ—Ç–µ –∑–º—ñ–Ω–∏—Ç–∏ —Ä—ñ–≤–µ–Ω—å –ª–æ–≥—É–≤–∞–Ω–Ω—è –¥–ª—è –æ–±—Ä–æ–±–Ω–∏–∫–∞. –ù–∞–ø—Ä–∏–∫–ª–∞–¥, —è–∫—â–æ –≤–∏ —Ö–æ—á–µ—Ç–µ –≤–∏–≤–æ–¥–∏—Ç–∏ —Ç—ñ–ª—å–∫–∏ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è INFO —ñ –≤–∏—â–µ, –≤—Å—Ç–∞–Ω–æ–≤—ñ—Ç—å `console_handler.setLevel(logging.INFO)`.

## –û–¥–Ω–æ—á–∞—Å–Ω–µ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è StreamHandler —Ç–∞ FileHandler

–í–∏ –º–æ–∂–µ—Ç–µ –æ–¥–Ω–æ—á–∞—Å–Ω–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ –æ–±–∏–¥–≤–∞ –æ–±—Ä–æ–±–Ω–∏–∫–∏ (`StreamHandler` —ñ `FileHandler`) —É –ª–æ–≥–µ—Ä—ñ –¥–ª—è –≤–∏–≤–æ–¥—É —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—ó —è–∫ –≤ –∫–æ–Ω—Å–æ–ª—å, —Ç–∞–∫ —ñ –≤ —Ñ–∞–π–ª. –û—Å—å –ø—Ä–∏–∫–ª–∞–¥:

```python
import logging

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –ª–æ–≥–µ—Ä–∞
logger = logging.getLogger(__name__)

# –ù–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è —Ä—ñ–≤–Ω—è –ª–æ–≥—É–≤–∞–Ω–Ω—è
logger.setLevel(logging.DEBUG)

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –æ–±—Ä–æ–±–Ω–∏–∫–∞ –¥–ª—è –≤–∏–≤–æ–¥—É –≤ stdout (–∫–æ–Ω—Å–æ–ª—å)
console_handler = logging.StreamHandler()

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –æ–±—Ä–æ–±–Ω–∏–∫–∞ –¥–ª—è –∑–∞–ø–∏—Å—É –≤ —Ñ–∞–π–ª
file_handler = logging.FileHandler('logfile.txt')

# –ù–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è —Ä—ñ–≤–Ω—è –ª–æ–≥—É–≤–∞–Ω–Ω—è –¥–ª—è –æ–±—Ä–æ–±–Ω–∏–∫—ñ–≤
console_handler.setLevel(logging.DEBUG)
file_handler.setLevel(logging.DEBUG)

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ñ–æ—Ä–º–∞—Ç–µ—Ä–∞ –¥–ª—è –æ–±—Ä–æ–±–Ω–∏–∫—ñ–≤
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')

# –ù–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è —Ñ–æ—Ä–º–∞—Ç–µ—Ä–∞ –¥–ª—è –æ–±—Ä–æ–±–Ω–∏–∫—ñ–≤
console_handler.setFormatter(formatter)
file_handler.setFormatter(formatter)

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –æ–±—Ä–æ–±–Ω–∏–∫—ñ–≤ –¥–æ –ª–æ–≥–µ—Ä–∞
logger.addHandler(console_handler)
logger.addHandler(file_handler)

# –õ–æ–≥—É–≤–∞–Ω–Ω—è –ø–æ–¥—ñ–π
logger.debug('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è DEBUG')
logger.info('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è INFO')
logger.warning('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è WARNING')
logger.error('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è ERROR')
logger.critical('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è CRITICAL')

```

–£ —Ü—å–æ–º—É –ø—Ä–∏–∫–ª–∞–¥—ñ –æ–±–∏–¥–≤–∞ –æ–±—Ä–æ–±–Ω–∏–∫–∏ (`console_handler` —ñ `file_handler`) –¥–æ–¥–∞—é—Ç—å—Å—è –¥–æ –ª–æ–≥–µ—Ä–∞, —ñ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –±—É–¥—É—Ç—å –≤–∏–≤–æ–¥–∏—Ç–∏—Å—è —è–∫ —É –∫–æ–Ω—Å–æ–ª—å, —Ç–∞–∫ —ñ —É —Ñ–∞–π–ª.

## –†—ñ–≤–Ω—ñ –ª–æ–≥—É–≤–∞–Ω–Ω—è

–†—ñ–≤–Ω—ñ –ª–æ–≥—É–≤–∞–Ω–Ω—è –≤–∏–∑–Ω–∞—á–∞—é—Ç—å –≤–∞–∂–ª–∏–≤—ñ—Å—Ç—å –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω—å –ª–æ–≥—ñ–≤ —ñ –¥–æ–∑–≤–æ–ª—è—é—Ç—å –≤–∞–º –∫–µ—Ä—É–≤–∞—Ç–∏ —Ç–∏–º, —è–∫—ñ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –±—É–¥—É—Ç—å –≤–∏–≤–æ–¥–∏—Ç–∏—Å—å –∞–±–æ –∑–∞–ø–∏—Å—É–≤–∞—Ç–∏—Å—å —É –ª–æ–≥-—Ñ–∞–π–ª. –£ –º–æ–¥—É–ª—ñ `logging` –≤ Python —ñ—Å–Ω—É—î –ø'—è—Ç—å —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∏—Ö —Ä—ñ–≤–Ω—ñ–≤ –ª–æ–≥—É–≤–∞–Ω–Ω—è:

1. **DEBUG (10):** –ù–∞–π–Ω–∏–∂—á–∏–π —Ä—ñ–≤–µ–Ω—å. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –≤—ñ–¥–ª–∞–¥–∫–∏. –ó–∞–∑–≤–∏—á–∞–π –≤–∏–≤–æ–¥–∏—Ç—å –¥–µ—Ç–∞–ª—ñ –ø—Ä–æ –≤–Ω—É—Ç—Ä—ñ—à–Ω—é —Ä–æ–±–æ—Ç—É –ø—Ä–æ–≥—Ä–∞–º–∏.
2. **INFO (20):** –Ü–Ω—Ñ–æ—Ä–º–∞—Ü—ñ–π–Ω–∏–π —Ä—ñ–≤–µ–Ω—å. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –ø–µ—Ä–µ–¥–∞—á—ñ —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—ó –ø—Ä–æ —Ö—ñ–¥ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –ø—Ä–æ–≥—Ä–∞–º–∏.
3. **WARNING (30):** –†—ñ–≤–µ–Ω—å –ø–æ–ø–µ—Ä–µ–¥–∂–µ–Ω—å. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –≤–∫–∞–∑—ñ–≤–∫–∏ –Ω–∞ –º–æ–∂–ª–∏–≤—ñ –ø—Ä–æ–±–ª–µ–º–∏, —è–∫—ñ –Ω–µ –ø—Ä–∏–∑–≤–æ–¥—è—Ç—å –¥–æ –≤—ñ–¥–º–æ–≤–∏ –ø—Ä–æ–≥—Ä–∞–º–∏, –∞–ª–µ –≤–∏–º–∞–≥–∞—é—Ç—å —É–≤–∞–≥–∏.
4. **ERROR (40):** –†—ñ–≤–µ–Ω—å –ø–æ–º–∏–ª–æ–∫. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –≤–∫–∞–∑—ñ–≤–∫–∏ –Ω–∞ —Å–µ—Ä–π–æ–∑–Ω—ñ –ø—Ä–æ–±–ª–µ–º–∏, —è–∫—ñ –ø—Ä–∏–∑–≤–æ–¥—è—Ç—å –¥–æ –≤—ñ–¥–º–æ–≤–∏ –ø—Ä–æ–≥—Ä–∞–º–∏.
5. **CRITICAL (50):** –ù–∞–π–≤–∏—â–∏–π —Ä—ñ–≤–µ–Ω—å. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –≤–∫–∞–∑—ñ–≤–∫–∏ –Ω–∞ –∫—Ä–∏—Ç–∏—á–Ω—ñ –ø–æ–º–∏–ª–∫–∏, —è–∫—ñ –º–æ–∂—É—Ç—å –ø—Ä–∏–∑–≤–µ—Å—Ç–∏ –¥–æ –∑—É–ø–∏–Ω–∫–∏ –ø—Ä–æ–≥—Ä–∞–º–∏.

–©–æ–± –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ —Ä—ñ–≤–Ω—ñ –ª–æ–≥—É–≤–∞–Ω–Ω—è –≤ –∫–æ–¥—ñ, –≤–∏ –º–æ–∂–µ—Ç–µ –≤—Å—Ç–∞–Ω–æ–≤–∏—Ç–∏ —Ä—ñ–≤–µ–Ω—å –¥–ª—è –ª–æ–≥–µ—Ä–∞ —á–∏ –æ–±—Ä–æ–±–Ω–∏–∫–∞. –ù–∞–ø—Ä–∏–∫–ª–∞–¥:

```python
import logging

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –ª–æ–≥–µ—Ä–∞
logger = logging.getLogger(__name__)

# –í—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è –ª–æ–≥—É–≤–∞–Ω–Ω—è –¥–ª—è –ª–æ–≥–µ—Ä–∞
logger.setLevel(logging.DEBUG)

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –æ–±—Ä–æ–±–Ω–∏–∫–∞ –¥–ª—è –≤–∏–≤–æ–¥—É –ª–æ–≥—ñ–≤ –Ω–∞ –∫–æ–Ω—Å–æ–ª—å
console_handler = logging.StreamHandler()

# –í—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è –ª–æ–≥—É–≤–∞–Ω–Ω—è –¥–ª—è –æ–±—Ä–æ–±–Ω–∏–∫–∞
console_handler.setLevel(logging.WARNING)

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –æ–±—Ä–æ–±–Ω–∏–∫–∞ –¥–æ –ª–æ–≥–µ—Ä–∞
logger.addHandler(console_handler)

# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è
logger.debug('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –¥–ª—è –≤—ñ–¥–ª–∞–¥–∫–∏')
logger.info('–¶–µ —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ–π–Ω–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è')
logger.warning('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è-–ø–æ–ø–µ—Ä–µ–¥–∂–µ–Ω–Ω—è')
logger.error('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –ø—Ä–æ –ø–æ–º–∏–ª–∫—É')
logger.critical('–¶–µ –∫—Ä–∏—Ç–∏—á–Ω–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è')

```

–£ —Ü—å–æ–º—É –ø—Ä–∏–∫–ª–∞–¥—ñ –ª–æ–≥–µ—Ä –≤—Å—Ç–∞–Ω–æ–≤–ª—é—î —Ä—ñ–≤–µ–Ω—å –ª–æ–≥—É–≤–∞–Ω–Ω—è –Ω–∞ `DEBUG`, —â–æ –¥–æ–∑–≤–æ–ª—è—î –ª–æ–≥—É–≤–∞—Ç–∏ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –≤—Å—ñ—Ö —Ä—ñ–≤–Ω—ñ–≤. –û–¥–Ω–∞–∫ –æ–±—Ä–æ–±–Ω–∏–∫ `console_handler` –≤—Å—Ç–∞–Ω–æ–≤–ª—é—î —Å–≤—ñ–π –≤–ª–∞—Å–Ω–∏–π —Ä—ñ–≤–µ–Ω—å –Ω–∞ `WARNING`, —Ç–æ–º—É –±—É–¥—É—Ç—å –≤–∏–≤–æ–¥–∏—Ç–∏—Å—å —Ç—ñ–ª—å–∫–∏ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—ñ–≤ `WARNING`, `ERROR` —ñ `CRITICAL`.

## –ú–æ–∂–ª–∏–≤–æ—Å—Ç—ñ —Ñ–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—è –ª–æ–≥–µ—Ä—É

–§–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—è –ª–æ–≥–µ—Ä–∞ –≤–∏–∑–Ω–∞—á–∞—î —Ç–µ, —è–∫—ñ —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ–π–Ω—ñ –µ–ª–µ–º–µ–Ω—Ç–∏ –≤–∫–ª—é—á–∞—é—Ç—å—Å—è —É –∫–æ–∂–Ω–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –ª–æ–≥—É. –í–∏ –º–æ–∂–µ—Ç–µ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ –∫–ª—é—á–æ–≤—ñ —Å–ª–æ–≤–∞ —Ñ–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—è –¥–ª—è –¥–æ–¥–∞–≤–∞–Ω–Ω—è —Ä—ñ–∑–Ω–∏—Ö –∞—Ç—Ä–∏–±—É—Ç—ñ–≤, —Ç–∞–∫–∏—Ö —è–∫ —á–∞—Å –ª–æ–≥—É–≤–∞–Ω–Ω—è, —Ä—ñ–≤–µ–Ω—å –ª–æ–≥—É–≤–∞–Ω–Ω—è, —Ç–µ–∫—Å—Ç –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ç–∞ —ñ–Ω—à—ñ. –û—Å—å –ø–æ—è—Å–Ω–µ–Ω–Ω—è –¥–µ—è–∫–∏—Ö –∑ –Ω–∏—Ö:

- **%(asctime)s**: –î–æ–¥–∞—î —á–∞—Å –ª–æ–≥—É–≤–∞–Ω–Ω—è —É —Ñ–æ—Ä–º–∞—Ç—ñ "—Ä—ñ–∫-–º—ñ—Å—è—Ü—å-–¥–µ–Ω—å –≥–æ–¥–∏–Ω–∞:—Ö–≤–∏–ª–∏–Ω–∞:—Å–µ–∫—É–Ω–¥–∞,–º—ñ–ª—ñ—Å–µ–∫—É–Ω–¥–∞".
- **%(levelname)s**: –î–æ–¥–∞—î —Ä—ñ–≤–µ–Ω—å –ª–æ–≥—É–≤–∞–Ω–Ω—è (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, DEBUG, INFO, WARNING, ERROR, CRITICAL).
- **%(message)s**: –î–æ–¥–∞—î —Ç–µ–∫—Å—Ç–æ–≤–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –ª–æ–≥—É.
- **%(filename)s**: –î–æ–¥–∞—î —ñ–º'—è —Ñ–∞–π–ª—É, –∑ —è–∫–æ–≥–æ –±—É–ª–æ –≤–∏–∫–ª–∏–∫–∞–Ω–æ –ª–æ–≥—É–≤–∞–Ω–Ω—è.
- **%(funcName)s**: –î–æ–¥–∞—î —ñ–º'—è —Ñ—É–Ω–∫—Ü—ñ—ó, –∑ —è–∫–æ—ó –±—É–ª–æ –≤–∏–∫–ª–∏–∫–∞–Ω–æ –ª–æ–≥—É–≤–∞–Ω–Ω—è.
- **%(lineno)d**: –î–æ–¥–∞—î –Ω–æ–º–µ—Ä —Ä—è–¥–∫–∞ —É —Ñ–∞–π–ª—ñ, –∑ —è–∫–æ–≥–æ –±—É–ª–æ –≤–∏–∫–ª–∏–∫–∞–Ω–æ –ª–æ–≥—É–≤–∞–Ω–Ω—è.
- **%(name)s**: –î–æ–¥–∞—î —ñ–º'—è –ª–æ–≥–µ—Ä–∞.

–í–∏ –º–æ–∂–µ—Ç–µ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ —Ü—ñ –∫–ª—é—á–æ–≤—ñ —Å–ª–æ–≤–∞ —É —Å–≤–æ—î–º—É —Ñ–æ—Ä–º–∞—Ç—ñ –¥–ª—è —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –≤–ª–∞—Å–Ω–æ–≥–æ —à–∞–±–ª–æ–Ω—É –ª–æ–≥—É–≤–∞–Ω–Ω—è. –û—Å—å –ø—Ä–∏–∫–ª–∞–¥, —è–∫ –≤—Å—Ç–∞–Ω–æ–≤–∏—Ç–∏ —Ñ–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—è –¥–ª—è –ª–æ–≥–µ—Ä–∞:

```python
import logging

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –ª–æ–≥–µ—Ä–∞
logger = logging.getLogger(__name__)

# –ù–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è —Ä—ñ–≤–Ω—è –ª–æ–≥—É–≤–∞–Ω–Ω—è
logger.setLevel(logging.DEBUG)

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –æ–±—Ä–æ–±–Ω–∏–∫–∞ –¥–ª—è –≤–∏–≤–æ–¥—É –≤ stdout (–∫–æ–Ω—Å–æ–ª—å)
console_handler = logging.StreamHandler()

# –ù–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è —Ä—ñ–≤–Ω—è –ª–æ–≥—É–≤–∞–Ω–Ω—è –¥–ª—è –æ–±—Ä–æ–±–Ω–∏–∫–∞
console_handler.setLevel(logging.DEBUG)

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ñ–æ—Ä–º–∞—Ç–µ—Ä–∞ –¥–ª—è –æ–±—Ä–æ–±–Ω–∏–∫–∞
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s - %(name)s - %(filename)s - %(lineno)d - %(funcName)s')

# –ù–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è —Ñ–æ—Ä–º–∞—Ç–µ—Ä–∞ –¥–ª—è –æ–±—Ä–æ–±–Ω–∏–∫–∞
console_handler.setFormatter(formatter)

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –æ–±—Ä–æ–±–Ω–∏–∫–∞ –¥–æ –ª–æ–≥–µ—Ä–∞
logger.addHandler(console_handler)

# –õ–æ–≥—É–≤–∞–Ω–Ω—è –ø–æ–¥—ñ–π
logger.debug('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è DEBUG')
logger.info('–¶–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è INFO')

```

–í —Ü—å–æ–º—É –ø—Ä–∏–∫–ª–∞–¥—ñ `%` –ø–µ—Ä–µ–¥ –∫–ª—é—á–æ–≤–∏–º —Å–ª–æ–≤–æ–º –≤–∫–∞–∑—É—î –Ω–∞ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è —Ü—å–æ–≥–æ –∫–ª—é—á–æ–≤–æ–≥–æ —Å–ª–æ–≤–∞ –¥–ª—è —Ñ–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—è. –ú–æ–∂–µ—Ç–µ –∑–º—ñ–Ω–∏—Ç–∏ –ø–æ—Ä—è–¥–æ–∫, –¥–æ–¥–∞–≤–∞—Ç–∏ –∞–±–æ –≤–∏–¥–∞–ª—è—Ç–∏ –∫–ª—é—á–æ–≤—ñ —Å–ª–æ–≤–∞, —â–æ–± –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—Ç–∏ –≤–∞—à–∏–º –ø–æ—Ç—Ä–µ–±–∞–º —Ñ–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—è –ª–æ–≥—É–≤–∞–Ω–Ω—è.

## –ü—Ä–∏–º—É—Å–æ–≤–µ –∑–∞–≤–µ—Ä—à–µ–Ω–Ω—è –ª–æ–≥—É–≤–∞–Ω–Ω—è

–ú–µ—Ç–æ–¥ `logging.shutdown()` –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –≤–∏–¥–∞–ª–µ–Ω–Ω—è –≤—Å—ñ—Ö –æ–±—Ä–æ–±–Ω–∏–∫—ñ–≤ –ª–æ–≥—É–≤–∞–Ω–Ω—è —Ç–∞ —ñ–Ω—à–∏—Ö —Ä–µ—Å—É—Ä—Å—ñ–≤, —è–∫—ñ –º–æ–∂—É—Ç—å –±—É—Ç–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω—ñ –º–æ–¥—É–ª–µ–º `logging`. –í–∏–∫–ª–∏–∫ —Ü—å–æ–≥–æ –º–µ—Ç–æ–¥—É –≤–∫–∞–∑—É—î, —â–æ –ª–æ–≥—É–≤–∞–Ω–Ω—è –≤ –ø—Ä–æ–≥—Ä–∞–º—ñ –∑–∞–≤–µ—Ä—à–∏–ª–æ —Å–≤–æ—é —Ä–æ–±–æ—Ç—É, —ñ –º–æ–∂–Ω–∞ –≤–∏–¥–∞–ª–∏—Ç–∏ –≤—Å—ñ –Ω–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è.

–¶–µ–π –º–µ—Ç–æ–¥ –∫–æ—Ä–∏—Å–Ω–∏–π —É –≤–∏–ø–∞–¥–∫–∞—Ö, –∫–æ–ª–∏ –ø—Ä–æ–≥—Ä–∞–º–∞ –∑–∞–≤–µ—Ä—à—É—î —Å–≤–æ—é —Ä–æ–±–æ—Ç—É, —ñ –≤–∏ —Ö–æ—á–µ—Ç–µ –∑–≤—ñ–ª—å–Ω–∏—Ç–∏ –≤—Å—ñ —Ä–µ—Å—É—Ä—Å–∏, –ø–æ–≤'—è–∑–∞–Ω—ñ —ñ–∑ —Å–∏—Å—Ç–µ–º–æ—é –ª–æ–≥—É–≤–∞–Ω–Ω—è. –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è:

```python
import logging

# –ù–∞–ª–∞—à—Ç–æ–≤—É—î–º–æ –ª–æ–≥–µ—Ä —Ç–∞ –æ–±—Ä–æ–±–Ω–∏–∫–∏
logger = logging.getLogger(__name__)
file_handler = logging.FileHandler('app.log')
logger.addHandler(file_handler)

# –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î–º–æ –ª–æ–≥–µ—Ä
logger.info('–¶–µ —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ–π–Ω–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è')

# –ó–∞–≤–µ—Ä—à—É—î–º–æ –ª–æ–≥—É–≤–∞–Ω–Ω—è —Ç–∞ –≤–∏–¥–∞–ª—è—î–º–æ —Ä–µ—Å—É—Ä—Å–∏
logging.shutdown()

```

–ü—Ä–∏ –≤–∏–∫–ª–∏–∫—É `logging.shutdown()`, –≤—Å—ñ –æ–±—Ä–æ–±–Ω–∏–∫–∏ –ª–æ–≥—ñ–≤ –±—É–¥—É—Ç—å –∑–∞–∫—Ä–∏—Ç—ñ, —ñ –±—É–¥–µ –≤–∏–∫–æ–Ω–∞–Ω–æ —ñ–Ω—à—ñ –¥—ñ—ó, —è–∫—ñ –º–æ–∂—É—Ç—å –±—É—Ç–∏ –Ω–µ–æ–±—Ö—ñ–¥–Ω—ñ –¥–ª—è –∑–∞–≤–µ—Ä—à–µ–Ω–Ω—è –ª–æ–≥—É–≤–∞–Ω–Ω—è.

## –°—Ç—Ä–æ–≥—ñ —Ç–≤–µ—Ä–¥–∂–µ–Ω–Ω—è assert

–í –º–æ–≤—ñ –ø—Ä–æ–≥—Ä–∞–º—É–≤–∞–Ω–Ω—è Python —Ç–≤–µ—Ä–¥–∂–µ–Ω–Ω—è `assert` –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –ø–µ—Ä–µ–≤—ñ—Ä–∫–∏ —ñ—Å—Ç–∏–Ω–Ω–æ—Å—Ç—ñ –≤–∏—Ä–∞–∑—É. –Ø–∫—â–æ –≤–∏—Ä–∞–∑ —î `False`, —Ç–æ–±—Ç–æ –Ω–µ –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—î –æ—á—ñ–∫—É–≤–∞–Ω–æ–º—É —Å—Ç–∞–Ω—É, —Ç–æ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –ø—Ä–æ–≥—Ä–∞–º–∏ –ø—Ä–∏–∑—É–ø–∏–Ω—è—î—Ç—å—Å—è, —ñ –≤–∏–∫–∏–¥–∞—î—Ç—å—Å—è –≤–∏–∫–ª—é—á–µ–Ω–Ω—è `AssertionError`.

–°–∏–Ω—Ç–∞–∫—Å–∏—Å —Ç–≤–µ—Ä–¥–∂–µ–Ω–Ω—è `assert` –≤–∏–≥–ª—è–¥–∞—î –Ω–∞—Å—Ç—É–ø–Ω–∏–º —á–∏–Ω–æ–º:

```python
assert –≤–∏—Ä–∞–∑, [–ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è_–ø—Ä–æ_–ø–æ–º–∏–ª–∫—É]
```

- `–≤–∏—Ä–∞–∑` - —Ü–µ —É–º–æ–≤–∞, —è–∫–∞ –ø–æ–≤–∏–Ω–Ω–∞ –±—É—Ç–∏ —ñ—Å—Ç–∏–Ω–Ω–æ—é.
- `–ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è_–ø—Ä–æ_–ø–æ–º–∏–ª–∫—É` (–Ω–µ–æ–±–æ–≤'—è–∑–∫–æ–≤–æ) - —Ü–µ —Ä—è–¥–æ–∫, —è–∫–∏–π –±—É–¥–µ –≤–∏–≤–µ–¥–µ–Ω–æ —É –≤–∏–ø–∞–¥–∫—É, —è–∫—â–æ —Ç–≤–µ—Ä–¥–∂–µ–Ω–Ω—è –≤–∏—è–≤–∏—Ç—å—Å—è —Ö–∏–±–Ω–∏–º. –¶–µ –¥–∞—î –º–æ–∂–ª–∏–≤—ñ—Å—Ç—å –∑–∞–∑–Ω–∞—á–∏—Ç–∏ –±—ñ–ª—å—à –¥–µ—Ç–∞–ª—å–Ω–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –ø—Ä–æ —Ç–µ, —â–æ –ø—ñ—à–ª–æ –Ω–µ —Ç–∞–∫.

–ü—Ä–∏–∫–ª–∞–¥–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è —Ç–≤–µ—Ä–¥–∂–µ–Ω–Ω—è `assert`:

```python
x = -5
assert x > 0, f"x=={x}, –∞–ª–µ –ø–æ–≤–∏–Ω–Ω–æ –±—É—Ç–∏ –¥–æ–¥–∞—Ç–Ω—ñ–º —á–∏—Å–ª–æ–º"
```

–£ —Ü—å–æ–º—É –ø—Ä–∏–∫–ª–∞–¥—ñ, —è–∫—â–æ –∑–Ω–∞—á–µ–Ω–Ω—è `x` –Ω–µ —î –¥–æ–¥–∞—Ç–Ω—ñ–º —á–∏—Å–ª–æ–º, –≤–∏–∫–æ–Ω–∞—î—Ç—å—Å—è —Ç–≤–µ—Ä–¥–∂–µ–Ω–Ω—è –ø—Ä–æ –ø–æ–º–∏–ª–∫—É —ñ –≤–∏–≤–µ–¥–µ—Ç—å—Å—è –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è `"–ø–æ–≤–∏–Ω–Ω–æ –±—É—Ç–∏ –¥–æ–¥–∞—Ç–Ω—ñ–º —á–∏—Å–ª–æ–º"`.

–¢–≤–µ—Ä–¥–∂–µ–Ω–Ω—è `assert` —á–∞—Å—Ç–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –ø–µ—Ä–µ–≤—ñ—Ä–∫–∏ –ø–æ–ø–µ—Ä–µ–¥–Ω—ñ—Ö —É–º–æ–≤, –¥–æ–ø–æ–º–∞–≥–∞—é—á–∏ –∑–∞–±–µ–∑–ø–µ—á–∏—Ç–∏ –∫–æ—Ä–µ–∫—Ç–Ω—ñ—Å—Ç—å –ø—Ä–æ–≥—Ä–∞–º–∏ —Ç–∞ —Ä–æ–∑—Ä–æ–±–ª—è—Ç–∏ —ó—ó –≤ —Ç–µ—Å—Ç–æ–≤–æ–º—É —Ä–µ–∂–∏–º—ñ.
