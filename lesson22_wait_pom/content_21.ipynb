## –ù–µ—è–≤–Ω–µ –æ—á—ñ–∫—É–∞–Ω–Ω—è

–£ `pytest` –∑–∞–∑–≤–∏—á–∞–π –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—Ç—å —Ñ—ñ–∫—Å—Ç—É—Ä–∏ –¥–ª—è —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –µ–∫–∑–µ–º–ø–ª—è—Ä—ñ–≤ Selenium —Ç–∞ —ñ–Ω—à–∏—Ö –Ω–µ–æ–±—Ö—ñ–¥–Ω–∏—Ö —Ä–µ—Å—É—Ä—Å—ñ–≤ –¥–ª—è —Ç–µ—Å—Ç—ñ–≤. –¶–µ –º–æ–∂–µ –±—É—Ç–∏ –∫–æ—Ä–∏—Å–Ω–æ –¥–ª—è –∞–≤—Ç–æ–º–∞—Ç–∏—á–Ω–æ–≥–æ –∑–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è —Ç–∞ –≤—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω—è –¥—Ä–∞–π–≤–µ—Ä—ñ–≤, —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –±—Ä–∞—É–∑–µ—Ä–Ω–∏—Ö —Å–µ—Å—ñ–π —Ç–æ—â–æ. 

–ù–µ—è–≤–Ω—ñ –æ—á—ñ–∫—É–≤–∞–Ω–Ω—è –º–æ–∂—É—Ç—å –±—É—Ç–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω—ñ –ø—Ä—è–º–æ –≤ –≤–∞—à–æ–º—É —Ç–µ—Å—Ç—ñ –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º –º–µ—Ç–æ–¥—ñ–≤ `implicitly_wait` –∞–±–æ `WebDriverWait`. –ù–∞–ø—Ä–∏–∫–ª–∞–¥:

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def test_example_with_implicit_wait(browser):
    browser.implicitly_wait(10)  # –ß–µ–∫–∞—Ç–∏ –Ω–µ –±—ñ–ª—å—à–µ 10 —Å–µ–∫—É–Ω–¥

    browser.get("<https://www.example.com>")

    # –ó–Ω–∞—Ö–æ–¥–∏–º–æ –µ–ª–µ–º–µ–Ω—Ç –Ω–∞ —Å—Ç–æ—Ä—ñ–Ω—Ü—ñ
    heading = browser.find_element_by_tag_name("h1")

    # –ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ, —á–∏ –≤—ñ—Ä–Ω–∏–π —Ç–µ–∫—Å—Ç –∑–∞–≥–æ–ª–æ–≤–∫—É
    assert heading.text == "Example Domain"

def test_example_with_explicit_wait(browser):
    browser.get("<https://www.example.com>")

    # –ß–µ–∫–∞—î–º–æ, –ø–æ–∫–∏ –∑–∞–≥–æ–ª–æ–≤–æ–∫ —Å—Ç–æ—Ä—ñ–Ω–∫–∏ –∑'—è–≤–∏—Ç—å—Å—è (–Ω–µ –±—ñ–ª—å—à–µ 10 —Å–µ–∫—É–Ω–¥)
    WebDriverWait(browser, 10).until(EC.presence_of_element_located((By.TAG_NAME, "h1")))

    # –ó–Ω–∞—Ö–æ–¥–∏–º–æ –µ–ª–µ–º–µ–Ω—Ç –Ω–∞ —Å—Ç–æ—Ä—ñ–Ω—Ü—ñ –ø—ñ—Å–ª—è –æ—á—ñ–∫—É–≤–∞–Ω–Ω—è
    heading = browser.find_element_by_tag_name("h1")

    # –ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ, —á–∏ –≤—ñ—Ä–Ω–∏–π —Ç–µ–∫—Å—Ç –∑–∞–≥–æ–ª–æ–≤–∫—É
    assert heading.text == "Example Domain"

```

## –Ø–≤–Ω–µ –æ—á—ñ–∫—É–≤–∞–Ω–Ω—è –∑ –ø–µ—Ä–µ–≤—ñ—Ä–∫–æ—é –Ω–∞—è–≤–Ω–æ—Å—Ç—ñ –∞–±–æ –≤—ñ–¥—Å—É—Ç–Ω–æ—Å—Ç—ñ

–î–ª—è –æ—á—ñ–∫—É–≤–∞–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç–∞ —Ç–∞ –ø–µ—Ä–µ–≤—ñ—Ä–∫–∏ –π–æ–≥–æ –Ω–∞—è–≤–Ω–æ—Å—Ç—ñ –∞–±–æ –≤—ñ–¥—Å—É—Ç–Ω–æ—Å—Ç—ñ –º–æ–∂–Ω–∞ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É Selenium —Ä–∞–∑–æ–º –∑—ñ —Å–ø–µ—Ü—ñ–∞–ª—å–Ω–∏–º–∏ —É–º–æ–≤–∞–º–∏ (Expected Conditions). –û—Å—å –∫—ñ–ª—å–∫–∞ –ø—Ä–∏–∫–ª–∞–¥—ñ–≤ –∫–æ–¥—É –Ω–∞ Python, —è–∫—ñ –¥–µ–º–æ–Ω—Å—Ç—Ä—É—é—Ç—å —Ü–µ–π –ø—ñ–¥—Ö—ñ–¥:

**–û—á—ñ–∫—É–≤–∞–Ω–Ω—è –Ω–∞—è–≤–Ω–æ—Å—Ç—ñ –µ–ª–µ–º–µ–Ω—Ç–∞**

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

def webelement(driver, xpath):
    try:
        element = WebDriverWait(driver, timeout=5).until(
            EC.presence_of_element_located((By.XPATH, xpath))
        )
        return element
    except TimeoutException:
        print("–ó–∞ –¥–∞–Ω–∏–π —á–∞—Å –µ–ª–µ–º–µ–Ω—Ç –Ω–µ –∑—è–≤–∏–≤—Å—è –Ω–∞ —Å—Ç–æ—Ä—ñ–Ω—Ü—ñ")

# –Ü–Ω—ñ—Ü—ñ–∞–ª—ñ–∑—É—î–º–æ –¥—Ä–∞–π–≤–µ—Ä –±—Ä–∞—É–∑–µ—Ä–∞
driver = webdriver.Chrome()
driver.get("https://www.example.com")
xpath = '//li[@id="ewuygeb"]'
element = webelement(driver, xpath)
element.click()
# –ó–∞–∫—Ä–∏—Ç—Ç—è –±—Ä–∞—É–∑–µ—Ä–∞
driver.quit()
```

```python
def element_is_not_here(driver, xpath):
    try:
        element = WebDriverWait(driver, timeout=5).until_not(
            EC.presence_of_element_located((By.XPATH, xpath))
        )
        return True
    except TimeoutException:
        print("–ï–ª–µ–º–µ–Ω—Ç –ø—Ä–æ–¥–æ–≤–∂—É—î –±—É—Ç–∏ –Ω–∞ —Å—Ç–æ—Ä—ñ–Ω—Ü—ñ –ø—ñ—Å–ª—è –æ—á—ñ–∫—É–≤–∞–Ω–æ–≥–æ —á–∞—Å—É")
        return False
```

[–ü–æ–≤–Ω–∏–π –ø–µ—Ä–µ–ª—ñ–∫ —è–≤–Ω–∏—Ö –æ—á—ñ–∫—É–≤–∞–Ω—å](https://www.selenium.dev/selenium/docs/api/py/webdriver_support/selenium.webdriver.support.expected_conditions.html)

## –î–æ–¥–∞—î–º–æ –≤–ª–∞—Å–Ω–µ —è–≤–Ω–µ –æ—á—ñ–∫—É–≤–∞–Ω–Ω—è

–í–∏ –º–æ–∂–µ—Ç–µ —Å—Ç–≤–æ—Ä–∏—Ç–∏ –≤–ª–∞—Å–Ω—ñ —É–º–æ–≤–∏ (Custom Conditions) –¥–ª—è –æ—á—ñ–∫—É–≤–∞–Ω—å (waits) —É Selenium, **—è–∫—â–æ –≤–±—É–¥–æ–≤–∞–Ω–∏—Ö –æ—á—ñ–∫—É–≤–∞–Ω—å –Ω–µ –≤–∏—Å—Ç–∞—á–∞—î** –¥–ª—è –≤–∞—à–æ–≥–æ –≤–∏–ø–∞–¥–∫—É –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è. –î–ª—è —Ü—å–æ–≥–æ –≤–∏ –º–æ–∂–µ—Ç–µ —Å—Ç–≤–æ—Ä–∏—Ç–∏ –ø—ñ–¥–∫–ª–∞—Å –∫–ª–∞—Å—É `ExpectedCondition` —ñ —Ä–µ–∞–ª—ñ–∑—É–≤–∞—Ç–∏ –º–µ—Ç–æ–¥ `__call__`, —è–∫–∏–π –ø–æ–≤–µ—Ä—Ç–∞—Ç–∏–º–µ `True`, —è–∫—â–æ –≤–∞—à–∞ —É–º–æ–≤–∞ –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è, –∞–±–æ `False`, —è–∫—â–æ –≤–æ–Ω–∞ –Ω–µ –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è –∞–±–æ —á–∞—Å –æ—á—ñ–∫—É–≤–∞–Ω–Ω—è –≤–∏—á–µ—Ä–ø–∞–Ω–æ.

–û—Å—å –ø—Ä–∏–∫–ª–∞–¥ —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –≤–ª–∞—Å–Ω–æ—ó —É–º–æ–≤–∏ –¥–ª—è –æ—á—ñ–∫—É–≤–∞–Ω—å:

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

class CustomCondition:
    def __init__(self, locator):
        self.locator = locator

    def __call__(self, driver):
        try:
            element = driver.find_element(*self.locator)  # –ó–Ω–∞—Ö–æ–¥–∏–º–æ –µ–ª–µ–º–µ–Ω—Ç
            return element.is_displayed()  # –ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ, —á–∏ –µ–ª–µ–º–µ–Ω—Ç –≤—ñ–¥–æ–±—Ä–∞–∂–∞—î—Ç—å—Å—è
        except:
            return False  # –Ø–∫—â–æ –µ–ª–µ–º–µ–Ω—Ç –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ –∞–±–æ –Ω–µ –≤—ñ–¥–æ–±—Ä–∞–∂–∞—î—Ç—å—Å—è, –ø–æ–≤–µ—Ä—Ç–∞—î–º–æ False

# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è
driver.get("<http://example.com>")
wait = WebDriverWait(driver, 10)

# –û—á—ñ–∫—É–≤–∞—Ç–∏, –ø–æ–∫–∏ –µ–ª–µ–º–µ–Ω—Ç —ñ–∑ –∑–∞–¥–∞–Ω–∏–º CSS —Å–µ–ª–µ–∫—Ç–æ—Ä–æ–º –Ω–µ –±—É–¥–µ –≤–∏–¥–∏–º–∏–º
element = wait.until(CustomCondition((By.CSS_SELECTOR, "div.some-class")))

# –ü–æ—Ç—ñ–º –ø—Ä–∞—Ü—é—î–º–æ –∑ –µ–ª–µ–º–µ–Ω—Ç–æ–º
element.click()

```

–£ —Ü—å–æ–º—É –ø—Ä–∏–∫–ª–∞–¥—ñ `CustomCondition` –ø–µ—Ä–µ–≤—ñ—Ä—è—î, —á–∏ –µ–ª–µ–º–µ–Ω—Ç —ñ–∑ –∑–∞–¥–∞–Ω–∏–º –ª–æ–∫–∞—Ç–æ—Ä–æ–º (CSS —Å–µ–ª–µ–∫—Ç–æ—Ä–æ–º) –≤—ñ–¥–æ–±—Ä–∞–∂–∞—î—Ç—å—Å—è –Ω–∞ —Å—Ç–æ—Ä—ñ–Ω—Ü—ñ. –Ø–∫—â–æ –µ–ª–µ–º–µ–Ω—Ç –≤—ñ–¥–æ–±—Ä–∞–∂–∞—î—Ç—å—Å—è, —É–º–æ–≤–∞ –≤–∏–∫–æ–Ω–∞—î—Ç—å—Å—è —ñ `wait.until()` –ø–æ–≤–µ—Ä–Ω–µ —Ü–µ–π –µ–ª–µ–º–µ–Ω—Ç. –Ø–∫—â–æ –µ–ª–µ–º–µ–Ω—Ç –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ –∞–±–æ –Ω–µ –≤—ñ–¥–æ–±—Ä–∞–∂–∞—î—Ç—å—Å—è –ø—Ä–æ—Ç—è–≥–æ–º —á–∞—Å—É –æ—á—ñ–∫—É–≤–∞–Ω–Ω—è, —É–º–æ–≤–∞ –ø–æ–≤–µ—Ä–Ω–µ `False`, —ñ –≤–∏–Ω–∏–∫–Ω–µ –≤–∏–Ω—è—Ç–æ–∫ `TimeoutException`.

# –î–µ–∫–æ–º–ø–æ–∑–∏—Ü—ñ—è —Ç–µ—Å—Ç–æ–≤–æ–≥–æ —Ñ—Ä–µ–π–º–≤–æ—Ä–∫—É

–î–µ–∫–æ–º–ø–æ–∑–∏—Ü—ñ—è —Ç–µ—Å—Ç–æ–≤–æ–≥–æ —Ñ—Ä–µ–π–º–≤–æ—Ä–∫—É –Ω–∞ –æ—Å–Ω–æ–≤—ñ Selenium —É Python –¥–æ–ø–æ–º–æ–∂–µ –≤–∞–º –∫—Ä–∞—â–µ –æ—Ä–≥–∞–Ω—ñ–∑—É–≤–∞—Ç–∏ –∫–æ–¥ —Ç–∞ –∑—Ä–æ–±–∏—Ç—å –π–æ–≥–æ –±—ñ–ª—å—à –∑—Ä–æ–∑—É–º—ñ–ª–∏–º, –ø—ñ–¥—Ç—Ä–∏–º—É–≤–∞–Ω–∏–º —ñ –ª–µ–≥–∫–æ —Ä–æ–∑—à–∏—Ä—é–≤–∞–Ω–∏–º. –û—Å—å –∫—ñ–ª—å–∫–∞ –æ—Å–Ω–æ–≤–Ω–∏—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ñ–≤ —Ç–∞ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü—ñ–π –∑ —ó—Ö —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—ó:

**–í–∏–∫–ª–∏–∫ –±—Ä–∞—É–∑–µ—Ä–∞ –±–µ–∑ –≤—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è –≤—ñ–∫–Ω–∞**

–ü–µ—Ä—à –Ω—ñ–∂ –º–∏ –∑–º–æ–∂–µ–º–æ –ø–µ—Ä–µ–π—Ç–∏ –¥–æ —Ç–µ—Å—Ç—É–≤–∞–Ω–Ω—è, —Å–ª—ñ–¥ —Ä–æ–∑–±–∏—Ç–∏ —Ñ—Ä–µ–π–º–≤–æ—Ä–∫ –Ω–∞ –æ–∫—Ä–µ–º—ñ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∏. –ù–∞–ø—Ä–∏–∫–ª–∞–¥, –º–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–º–æ –¥—Ä–∞–π–≤–µ—Ä, —É –≤—Å—ñ—Ö —Ç–µ—Å—Ç–∞—Ö:

C—Ç–≤–æ—Ä–∏–º–æ —Ñ–∞–π–ª `get_browser.py`

```python
def firefox(debug=False):
    # –Ü–Ω—ñ—Ü—ñ–∞–ª—ñ–∑—É—î–º–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏ –¥–ª—è –≤–µ–±-–¥—Ä–∞–π–≤–µ—Ä–∞ Firefox
    options = FirefoxOptions()
    options.add_argument("--headless")  # –ó–∞–ø—É—Å–∫ —É –±–µ–∑–≥–æ–ª–æ–≤–Ω–æ–º—É —Ä–µ–∂–∏–º—ñ
    # (–±–µ–∑ –≤—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è –≤—ñ–∫–Ω–∞ –±—Ä–∞—É–∑–µ—Ä–∞)

    # –í–∏–±–∏—Ä–∞—î–º–æ, —á–∏ –ø–æ—Ç—Ä—ñ–±–Ω–æ –∑–∞–ø—É—Å–∫–∞—Ç–∏ –¥—Ä–∞–π–≤–µ—Ä —É —Ä–µ–∂–∏–º—ñ –Ω–∞–ª–∞–≥–æ–¥–∂–µ–Ω–Ω—è (debug mode)
    driver = webdriver.Firefox() if debug else \
             webdriver.Firefox(options=options)  
    # –í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è headless, —è–∫—â–æ –Ω–µ –≤–∫–ª—é—á–µ–Ω–æ –Ω–∞–ª–∞–≥–æ–¥–∂–µ–Ω–Ω—è

    # –ú–∞–∫—Å–∏–º—ñ–∑—É—î–º–æ –≤—ñ–∫–Ω–æ –±—Ä–∞—É–∑–µ—Ä–∞ (—è–∫—â–æ –≤—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è –Ω–µ –≤–∏–º–∫–Ω–µ–Ω–æ)
    driver.maximize_window()

    # –ü–æ–≤–µ—Ä—Ç–∞—î–º–æ –æ–±'—î–∫—Ç –¥—Ä–∞–π–≤–µ—Ä–∞
    return driver
```

–¶–µ –¥–æ–∑–≤–æ–ª–∏—Ç—å –∑–∞–ø—É—Å–∫–∞—Ç–∏ –±—Ä–∞—É–∑–µ—Ä Firefox –±–µ–∑ –≤—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è –≤—ñ–∫–Ω–∞, —â–æ –ø—Ä–∏—Å–∫–æ—Ä–∏—Ç—å —Ç–µ—Å—Ç–∏, –∞–ª–µ —è–∫—â–æ –Ω–∞–º –ø–æ—Ç—Ä—ñ–±–Ω–æ –ø–æ–±–∞—á–∏—Ç–∏, —â–æ –≤—ñ–¥–±—É–≤–∞—î—Ç—å—Å—è –º–∏ –º–æ–∂–µ–º–æ –ø—Ä–æ—Å—Ç–æ –≤–∏–∫–ª–∏–∫–∞—Ç–∏ `firefox(True)` –ø—ñ–¥ —á–∞—Å –≤–∏–∫–æ–Ω–∞–Ω–Ω—è —Ç–µ—Å—Ç—ñ–≤.

**–í–∏–Ω—ñ—Å —Ä–æ–±–æ—Ç–∏ –∑ –≤–µ–±-–µ–ª–µ–º–µ–Ω—Ç–∞–º–∏ –≤ –æ–∫—Ä–µ–º—É —Ñ—É–Ω–∫—Ü—ñ—é**

—Å—Ç–≤–æ—Ä–∏–º–æ —Ñ–∞–π–ª `processing_element.py`

```python
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

def find_element(driver, xpath):
    """–§—É–Ω–∫—Ü—ñ—è –¥–ª—è –æ—á—ñ–∫—É–≤–∞–Ω–Ω—è –Ω–∞—è–≤–Ω–æ—Å—Ç—ñ –µ–ª–µ–º–µ–Ω—Ç–∞ –Ω–∞ —Å—Ç–æ—Ä—ñ–Ω—Ü—ñ"""
    try:
        element = WebDriverWait(driver, timeout=5).until(
            EC.presence_of_element_located((By.XPATH, xpath))
        )
        return element
    except TimeoutException:
        raise NoSuchElementException("–ó–∞ –¥–∞–Ω–∏–π —á–∞—Å –µ–ª–µ–º–µ–Ω—Ç –Ω–µ –∑—è–≤–∏–≤—Å—è –Ω–∞ —Å—Ç–æ—Ä—ñ–Ω—Ü—ñ")

def click_element(driver, xpath):
    """–§—É–Ω–∫—Ü—ñ—è –¥–ª—è –∫–ª—ñ–∫—É –Ω–∞ –µ–ª–µ–º–µ–Ω—Ç"""
    element = find_element(driver, by, value)
    element.click()

def fill_input(driver, xpath, text):
    """–§—É–Ω–∫—Ü—ñ—è –¥–ª—è –∑–∞–ø–æ–≤–Ω–µ–Ω–Ω—è –ø–æ–ª—è –≤–≤–æ–¥—É —Ç–µ–∫—Å—Ç–æ–º"""
    element = find_element(driver, xpath)
    element.clear()
    element.send_keys(text)
```

## –Ü–º–ø–æ—Ä—Ç –∑–∞–ª–µ–∂–Ω–æ—Å—Ç–µ–π
